# INTRODUCTION
**What is Programming?**<br>
Programming is the process of giving instructions to a computer to perform a specific task.

A computer cannot think or decide on its own. It works only according to the instructions given by humans. **These instructions are written using a programming language and are called a program.**

Example:
Asking a computer to calculate marks
Asking a computer to store student data
Asking a computer to display a result

**Types of Programing Language**<br>
**1. High-Level Programming Language**<br>
A high-level programming language is designed to be easy for humans to read and write. It hides hardware details and uses simple, English-like syntax. Example: Python, Java, C#

  **2. Low-Level Programming Language**<br>
A low-level programming language is close to machine hardware and difficult for humans to understand. It provides more control over system resources. Example: Assembly language,

---



**What is a General-Purpose Programming Language**<br>
A general-purpose programming language is used to develop different types of applications such as web apps, desktop software, data analysis, and automation. Example: Python, C++, Java

**What is an Interpreted Language**<br>
An interpreted language executes code line by line using an interpreter. It does not require prior compilation, making debugging easier. Example: Python, JavaScript

**What is a Compiled Language** <br>
A compiled language converts the entire source code into machine code before execution using a compiler. This generally results in faster execution. Example: C, C++


**What is Python?**

Python is a high-level, interpreted, and general-purpose programming language.
It was created by Guido van Rossum and released in 1991.
Python is known for its simple syntax, readability, and wide range of applications.

Python allows programmers to write fewer lines of code compared to many other programming languages, making it easy to learn and use.

## Programming Development Tools Overview
### Compiler

A **compiler** is a program that translates the entire source code into machine code **before execution**.
**Example:** GCC (C/C++), javac (Java)


### Interpreter

An **interpreter** executes code **line by line**, without converting the whole program at once.
**Example:** Python, JavaScript

### IDE

An **IDE (Integrated Development Environment)** is a software tool that provides a complete environment to write, run, and debug programs.
**Examples (for Python):**

* PyCharm
* Visual Studio Code
* Spyder

### IDLE

**IDLE (Integrated Development and Learning Environment)** is the **default IDE provided with Python**, mainly used by beginners.
**Features:** Editor, Python shell, basic debugging

### Other IDEs / IDLE-like Tools for Python

* PyCharm
* Visual Studio Code
* Jupyter Notebook
* Thonny
* Google Collab
