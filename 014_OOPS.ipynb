### What is Object-Oriented Programming (OOP)?

OOP is a way of organizing your code around **data (attributes)** and **functions (methods)** that operate on that data, rather than just a sequence of functions and logic. Think of it like building with LEGO bricks, where each brick (object) has its own properties and actions.

### 1. Classes and Objects

*   **Class:** A blueprint or a template for creating objects. It defines what properties (attributes) an object will have and what actions (methods) it can perform.
*   **Object:** An instance of a class. It's a real entity created from the blueprint. You can create many objects from a single class.

```python
# Define a simple class named 'Dog'
class Dog:
    pass # 'pass' means this class doesn't do anything yet

# Create an object (an instance) of the Dog class
my_dog = Dog()
your_dog = Dog()

print(f"Type of my_dog: {type(my_dog)}")
print(f"Type of your_dog: {type(your_dog)}")
print(f"Are my_dog and your_dog the same object? {my_dog is your_dog}")
```

### 2. Attributes

*   **Attributes:** These are variables that belong to an object. They store data about the object. Think of them as characteristics of an object (e.g., a dog's name, age, breed).

```python
class Dog:
    pass

# Create a dog object
my_dog = Dog()

# Assign attributes to the object
my_dog.name = "Buddy"
my_dog.age = 3
my_dog.breed = "Golden Retriever"

# Access and print the attributes
print(f"My dog's name is {my_dog.name}.")
my_dog.name = "Huddy"
print(f"My dog's name is {my_dog.name}.")
print(f"My dog is {my_dog.age} years old.")
print(f"My dog is a {my_dog.breed}.")

# You can also create another dog with different attributes
your_dog = Dog()
your_dog.name = "Lucy"
your_dog.age = 5

print(f"\nYour dog's name is {your_dog.name}.")
```

### 3. Methods

*   **Methods:** These are functions that belong to an object. They define the actions or behaviors an object can perform (e.g., a dog can bark, eat, sleep).
*   Methods always have `self` as their first parameter, which refers to the instance of the object itself.

```python
class Dog:
    def bark(self):
        print("Woof! Woof!")

    def eat(self, food):
        print(f"The dog is eating {food}.")

# Create a dog object
my_dog = Dog()

# Call the methods on the object
my_dog.bark()
my_dog.eat("kibble")
```

### 4. The `__init__` Method (Constructor)

*   A constructor is a function within a class that initializes an object.
*   The `__init__` method is a special method in Python classes.
*   It's automatically called whenever a new object (instance) of the class is created.
*   Its main purpose is to initialize the object's attributes (set their starting values) when the object is created.


```python
class Dog:
    # The __init__ method takes 'self' and other parameters (name, age, breed)
    def __init__(self, name, age, breed):
        self.name = name     # Initialize the 'name' attribute
        self.age = age       # Initialize the 'age' attribute
        self.breed = breed   # Initialize the 'breed' attribute

    def bark(self):
        print(f"{self.name} says Woof! Woof!")

    def description(self):
        return f"{self.name} is a {self.age}-year-old {self.breed}."

# Create dog objects, passing initial values to the __init__ method
dog1 = Dog("Buddy", 3, "Golden Retriever")
dog2 = Dog("Lucy", 5, "Labrador")

print(dog1.description())
dog1.bark()

print(dog2.description())
dog2.bark()
```

2.  **Parameterized Constructor:** A constructor that accepts additional arguments to initialize the object's attributes with specific values provided at creation.
3.  **Constructor with Default Arguments:** A parameterized constructor where some (or all) parameters have default values, making them optional during object creation.

### getattr`, `setattr`, and `delattr`

These functions are useful when you need to interact with object attributes dynamically, for example, when attribute names are stored as strings or determined at runtime. They provide a more programmatic way to manage attributes compared to direct dot notation.

*   **`getattr(object, name[, default])`**: Used to retrieve the value of an attribute given its name as a string. You can also provide a default value to avoid `AttributeError` if the attribute doesn't exist.
*   **`setattr(object, name, value)`**: Used to set the value of an attribute given its name as a string and the new value. This can also be used to add new attributes to an object dynamically.
*   **`delattr(object, name)`**: Used to delete an attribute from an object given its name as a string.

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def display_product(self):
        print(f"Product: {self.name}, Price: ${self.price}, Stock: {self.stock}")

# Create a Product object
my_product = Product("Laptop", 1200.00, 50)
my_product.display_product()

print("\n--- Using getattr() ---")
# Accessing attributes using getattr()
product_name = getattr(my_product, "name")
product_price = getattr(my_product, "price")
print(f"Accessed Name: {product_name}")
print(f"Accessed Price: {product_price}")

# Trying to access a non-existent attribute with a default value
product_description = getattr(my_product, "description", "No description available")
print(f"Accessed Description: {product_description}")


print("\n--- Using setattr() ---")
# Updating an existing attribute using setattr()
setattr(my_product, "price", 1150.00) # Update price
setattr(my_product, "stock", 45)     # Update stock
print("After updating price and stock:")
my_product.display_product()

# Adding a new attribute using setattr()
setattr(my_product, "category", "Electronics")
print(f"Added new attribute 'category': {my_product.category}")

print("\n--- Using delattr() ---")
# Deleting an attribute using delattr()
print(f"Before deleting category: {hasattr(my_product, 'category')}")
delattr(my_product, "category")
print(f"After deleting category: {hasattr(my_product, 'category')}")
```

### What is an Instance Variable?

An **instance variable** is a variable that belongs to an instance (object) of a class. Each object created from a class will have its own independent copy of these variables.


### 1. Creating Instance Variables Using the Constructor (`__init__` method)

This is the most common and recommended way to initialize instance variables. The `__init__` method is automatically called when a new object is created, allowing you to set up its initial state.

All variables assigned to `self` within the `__init__` method become instance variables.

```python
class Person:
    def __init__(self, name, age): # The constructor method
        self.name = name # Instance variable 'name'
        self.age = age   # Instance variable 'age'
        self.species = "Human" # Another instance variable, initialized with a default value

    def display_info(self):
        print(f"{self.name} is {self.age} years old and is a {self.species}.")

# Creating objects (instances)
person1 = Person("Alice", 30)
person2 = Person("Bob", 25)

# Accessing instance variables directly
print(f"Person 1's name: {person1.name}")
print(f"Person 2's age: {person2.age}")

# Calling an instance method that uses instance variables
person1.display_info()
person2.display_info()
```

### 2. Creating Instance Variables Outside the Class (Dynamically)

Python allows you to add new attributes to an object dynamically after it has been created, using the dot (`.`) notation directly on the object. These are also instance variables, unique to that specific object.

While Python allows this flexibility, it's generally good practice to define all expected instance variables in the `__init__` method for clarity and maintainability. Dynamic addition can make code harder to follow, but it's useful in certain situations (e.g., adding temporary flags).

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

# Create a book object
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book("1984", "George Orwell")

# Dynamically add a new instance variable to book1
book1.publisher = "Scribner"
print(f"Book 1 title: {book1.title}, Publisher: {book1.publisher}")


# You can also modify existing instance variables outside the class
book1.title = "The Great Gatsby (Revised Edition)"
print(f"Updated Book 1 title: {book1.title}")
```

### 3. Creating Instance Variables Using an Instance Method

You can also create new instance variables within any instance method (not just `__init__`). However, this is less common for core attributes and typically used for attributes that are calculated or set later based on some action.

**Caution:** If an instance variable is only created in a specific method, it won't exist for objects that don't call that method, which can lead to `AttributeError` if not handled carefully.

```python
class Student:
    def __init__(self, name, marks):
        # existing instance variables
        self.name = name
        self.marks = marks

    # instance method that creates a NEW instance variable
    def calculate_grade(self):
        if self.marks >= 90:
            self.grade = "A"
        elif self.marks >= 75:
            self.grade = "B"
        else:
            self.grade = "C"

        print("Grade created for", self.name)


# create object
s1 = Student("Rohit", 82)

# calling method (this will CREATE new instance variable 'grade')
s1.calculate_grade()

# accessing new instance variable
print("Name:", s1.name)
print("Marks:", s1.marks)
print("Grade:", s1.grade)
```

### 1. Instance Method

*   **Definition:** An instance method is the most common type of method in a Python class. It operates on an instance (object) of the class and can access and modify the instance's attributes.
*   **First Argument:** Always takes `self` as its first parameter, which refers to the specific instance (object) on which the method is called.
*   **Purpose:** Used for actions that are specific to a particular object, often involving the object's state.
*   **Decorator:** None (no special decorator is needed).

```python
class Dog:
    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute

    def bark(self): # This is an instance method
        print(f"{self.name} says Woof! Woof!")

    def celebrate_birthday(self): # This is an instance method
        self.age += 1 # Modifies the instance's state
        print(f"{self.name} is now {self.age} years old! Happy birthday!")

# Create instances (objects)
my_dog = Dog("Buddy", 3)
your_dog = Dog("Lucy", 5)

# Call instance methods on specific instances
my_dog.bark()
my_dog.celebrate_birthday()

your_dog.bark()
print(f"Buddy's age: {my_dog.age}, Lucy's age: {your_dog.age}")
```

### What is a Class Variable?

A **class variable** is a variable that is owned by the class itself, rather than by any specific instance of the class. This means that all instances of the class share the same class variable.

Changes made to a class variable by any  will be reflected across all instances.

*   **Where to define:** Class variables are defined inside the class, but outside of any instance method (like `__init__`).

```python
class Robot:
    # This is a Class Variable
    # It's shared by all instances of the Robot class
    population = 0
    all_robots = [] # Another class variable to keep track of all robots

    def __init__(self, name, model):
        self.name = name # This is an Instance Variable
        self.model = model # This is an Instance Variable
        Robot.population += 1 # Increment the class variable
        Robot.all_robots.append(self.name) # Add instance name to class list
        print(f"Robot {self.name} ({self.model}) created.")

    def introduce(self):
        print(f"Hello, my name is {self.name} and I am a {self.model} model.")

    def get_population(self):
        # Accessing class variable via instance
        print(f"There are {self.population} robots in total (accessed via instance).")


# --- Creating Robot instances ---
print("--- Creating Robots ---")
robot1 = Robot("R2D2", "Astromech")
robot2 = Robot("C3PO", "Protocol")

# --- Accessing Class Variables ---
print("\n--- Accessing Class Variables ---")
# Accessing population directly via the class
print(f"Total robot population: {Robot.population} (accessed via class).")

# Accessing all_robots list directly via the class
print(f"All robot names: {Robot.all_robots}")

# Accessing population via an instance (shows the shared value)
robot1.get_population()
robot2.get_population()

# --- Modifying a Class Variable ---
print("\n--- Modifying Class Variable ---")
Robot.population = 10 # Directly change the class variable
print(f"New total robot population: {Robot.population}")
print(f"Robot1's population view: {robot1.population}") # Instance now reflects the change

# --- Demonstrating Instance vs. Class Variable ---
print("\n--- Instance vs. Class Variable ---")
robot3 = Robot("BB8", "Droid") # population will be 11 now
print(f"Robot3's name (instance variable): {robot3.name}")
print(f"Total population after creating robot3: {Robot.population}")
```

### 2. Class Method

*   **Definition:** A class method is a method that is bound to the class itself, not to an instance of the class. It can access or modify the class's state (class variables) but not directly the instance's state.
*   **First Argument:** Takes `cls` as its first argument, which refers to the class object itself.
*   **Purpose:** Commonly used for factory methods (alternative constructors that return an instance of the class) or to modify class-wide properties.
*   **Decorator:** Uses the `@classmethod` decorator.

```python
class Car:
    total_cars_produced = 0 # This is a class variable

    def __init__(self, make, model):
        self.make = make
        self.model = model
        Car.total_cars_produced += 1 # Modifies the class variable

    @classmethod
    def get_total_cars_produced(cls): # This is a class method
        return f"Total cars produced: {cls.total_cars_produced}"

    @classmethod
    def from_string(cls, car_string): # This is a factory method (class method)
        make, model = car_string.split('-')
        return cls(make, model) # Returns an instance of the class

    def display_info(self):
        return f"{self.make} {self.model}"

# Create instances
car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")

# Call the class method using the class name
print(Car.get_total_cars_produced())

# Use the factory method to create a new car
car3 = Car.from_string("Ford-Mustang")
print(f"New car from string: {car3.display_info()}")
print(Car.get_total_cars_produced())
```

### 3. Static Method

*   **Definition:** A static method belongs to the class but does not operate on the instance or the class itself. It's essentially a regular function logically grouped within a class.
*   **First Argument:** It doesn't implicitly receive `self` or `cls` as its first argument.
*   **Purpose:** Used for utility or helper functions that don't depend on the object's state or the class's state.
*   **Decorator:** Uses the `@staticmethod` decorator.

```python
class Calculator:
    def __init__(self, name):
        self.name = name

    @staticmethod
    def add(x, y): # This is a static method
        return x + y

    @staticmethod
    def subtract(x, y): # This is a static method
        return x - y

    def describe(self):
        print(f"This is {self.name}, capable of basic operations.")

# Call static methods directly using the class name
result_add = Calculator.add(10, 5)
result_subtract = Calculator.subtract(10, 5)

print(f"Result of addition: {result_add}")
print(f"Result of subtraction: {result_subtract}")

# Static methods can also be called via an instance, but they don't use the instance
my_calc = Calculator("My Super Calculator")
my_calc.describe()
result_add_via_instance = my_calc.add(20, 7)
print(f"Result of addition via instance: {result_add_via_instance}")
```

### Summary of Method Types:

| Feature            | Instance Method      | Class Method            | Static Method                 |
| :----------------- | :------------------- | :---------------------- | :---------------------------- |
| **First Argument** | `self` (instance)    | `cls` (class)           | None (`self` or `cls` not used) |
| **Accesses**       | Instance data & Class data | Class data only         | Neither instance nor class data |
| **Usage**          | Operations on object state | Factory methods, modify class state | Utility functions, helper functions |
| **Decorator**      | None                 | `@classmethod`          | `@staticmethod`               |

In essence:
*   Use **instance methods** when you need to access or modify the state of a specific object.
*   Use **class methods** when you need to access or modify the state of the class itself, or to create alternative constructors.
*   Use **static methods** when the method logically belongs to the class but doesn't need to interact with the class's or an instance's state.

### 1. Car Manufacturing System

This example models a car manufacturing system, demonstrating how different types of cars can inherit common properties and behaviors while also having their own unique characteristics. It highlights:

*   **Inheritance:** `ElectricCar` and `GasCar` inherit from a base `Car` class.
*   **Polymorphism:** The `start_engine` method behaves differently for electric and gas cars.
*   **Class Variables:** A `total_cars_produced` tracks all cars made.
*   **Encapsulation:** Attributes like `_fuel_level` are managed internally.

```python
class Car:
    total_cars_produced = 0 # Class variable to count all cars

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        Car.total_cars_produced += 1

    def display_info(self):
        return f"{self.year} {self.make} {self.model}"

    def start_engine(self):
        raise NotImplementedError("Subclasses must implement this method")

class GasCar(Car):
    def __init__(self, make, model, year, fuel_capacity):
        super().__init__(make, model, year)
        self._fuel_level = 0 # Protected attribute
        self.fuel_capacity = fuel_capacity

    def start_engine(self):
        if self._fuel_level > 0:
            return f"{self.display_info()}: Vroom! Engine started."
        else:
            return f"{self.display_info()}: Out of gas! Can't start."

    def refuel(self, amount):
        if amount > 0:
            self._fuel_level = min(self._fuel_level + amount, self.fuel_capacity)
            print(f"{self.make} {self.model} refueled. Fuel level: {self._fuel_level}/{self.fuel_capacity}")
        else:
            print("Refuel amount must be positive.")

class ElectricCar(Car):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self._charge_level = 100 # Initially fully charged
        self.battery_capacity = battery_capacity

    def start_engine(self):
        if self._charge_level > 10:
            return f"{self.display_info()}: Whirr... Electric motor engaged."
        else:
            return f"{self.display_info()}: Low charge! Please charge battery."

    def charge(self, percentage):
        if 0 < percentage <= 100:
            self._charge_level = min(self._charge_level + percentage, 100)
            print(f"{self.make} {self.model} charged. Charge level: {self._charge_level}%")
        else:
            print("Charge percentage must be between 1 and 100.")

# Create objects
my_gas_car = GasCar("Ford", "Mustang", 2022, 15)
my_electric_car = ElectricCar("Tesla", "Model 3", 2023, 75)

print(my_gas_car.start_engine())
my_gas_car.refuel(10)
print(my_gas_car.start_engine())

print("\n")
print(my_electric_car.start_engine())
my_electric_car.charge(5)
print(my_electric_car.start_engine())

print(f"\nTotal cars produced: {Car.total_cars_produced}")
```

### Simplified Car Manufacturing System (for Beginners)

This version of the car manufacturing system uses a single `Car` class to represent both gas and electric vehicles. Instead of inheritance, it uses an `engine_type` attribute to distinguish between them and applies conditional logic within the methods to provide the correct behavior for each type. This approach is often easier to grasp for those new to object-oriented programming.

We will still track the total number of cars produced using a class variable, as it's a simple and useful concept.

```python
class Car:
    total_cars_produced = 0 # Class variable to count all cars

    def __init__(self, make, model, year, engine_type, fuel_capacity=None, battery_capacity=None):
        self.make = make
        self.model = model
        self.year = year
        self.engine_type = engine_type.lower() # 'gas' or 'electric'
        Car.total_cars_produced += 1

        if self.engine_type == 'gas':
            self.fuel_capacity = fuel_capacity
            self._fuel_level = 0 # Initial fuel level
            self.battery_capacity = None # Not applicable for gas cars
            self._charge_level = None # Not applicable
        elif self.engine_type == 'electric':
            self.battery_capacity = battery_capacity
            self._charge_level = 100 # Initially fully charged
            self.fuel_capacity = None # Not applicable for electric cars
            self._fuel_level = None # Not applicable
        else:
            print("Invalid engine type. Please choose 'gas' or 'electric'.")
            # Optionally, raise an error or handle invalid type

    def display_info(self):
        return f"{self.year} {self.make} {self.model}"

    def start_engine(self):
        if self.engine_type == 'gas':
            if self._fuel_level > 0:
                return f"{self.display_info()}: Vroom! Engine started."
            else:
                return f"{self.display_info()}: Out of gas! Can't start."
        elif self.engine_type == 'electric':
            if self._charge_level > 10:
                return f"{self.display_info()}: Whirr... Electric motor engaged."
            else:
                return f"{self.display_info()}: Low charge! Please charge battery."
        else:
            return "Cannot start engine due to invalid engine type."

    def refuel(self, amount):
        if self.engine_type == 'gas':
            if amount > 0:
                self._fuel_level = min(self._fuel_level + amount, self.fuel_capacity)
                print(f"{self.make} {self.model} refueled. Fuel level: {self._fuel_level}/{self.fuel_capacity}")
            else:
                print("Refuel amount must be positive.")
        else:
            print(f"{self.make} {self.model} is an electric car, cannot refuel.")

    def charge(self, percentage):
        if self.engine_type == 'electric':
            if 0 < percentage <= 100:
                self._charge_level = min(self._charge_level + percentage, 100)
                print(f"{self.make} {self.model} charged. Charge level: {self._charge_level}%")
            else:
                print("Charge percentage must be between 1 and 100.")
        else:
            print(f"{self.make} {self.model} is a gas car, cannot charge.")

# --- Creating Car Objects ---
# Create a gas car
my_gas_car = Car("Ford", "Mustang", 2022, "gas", fuel_capacity=15)

# Create an electric car
my_electric_car = Car("Tesla", "Model 3", 2023, "electric", battery_capacity=75)

print("\n--- Gas Car Actions ---")
print(my_gas_car.start_engine())
my_gas_car.refuel(10)
print(my_gas_car.start_engine())

print("\n--- Electric Car Actions ---")
print(my_electric_car.start_engine())
my_electric_car.charge(5)
print(my_electric_car.start_engine())

print(f"\nTotal cars produced: {Car.total_cars_produced}")

# Demonstrating incorrect actions (for illustration)
print("\n--- Incorrect Actions (for illustration) ---")
my_gas_car.charge(20) # Trying to charge a gas car
my_electric_car.refuel(5) # Trying to refuel an electric car
```

### This example models a very simple library system using basic class and object concepts.

Basic Object Interaction: A Borrower object interacts with a single Book object through the Library class.
State Management: The is_borrowed attribute is used to check whether the book is available or not.<br>
Methods for Actions: Simple methods like add_book, add_borrower, borrow_book, and return_book are used to perform operations.<br>
Basic Condition Handling: Uses simple if-else statements to prevent borrowing an already borrowed book or returning an already available book.

```python
class Book:
    def __init__(self, title):
        self.title = title
        self.is_borrowed = False


class Borrower:
    def __init__(self, name):
        self.name = name


class Library:
    def __init__(self):
        self.book = None
        self.borrower = None

    def add_book(self, book):
        self.book = book
        print("Book added:", book.title)

    def add_borrower(self, borrower):
        self.borrower = borrower
        print("Borrower added:", borrower.name)

    def borrow_book(self):
        if self.book.is_borrowed == False:
            self.book.is_borrowed = True
            print(self.borrower.name, "borrowed", self.book.title)
        else:
            print("Book already borrowed")

    def return_book(self):
        if self.book.is_borrowed == True:
            self.book.is_borrowed = False
            print(self.borrower.name, "returned", self.book.title)
        else:
            print("Book is already available")


  # Create objects
book1 = Book("Python Basics")
person1 = Borrower("Rohit")

lib = Library()

# Add book and borrower
lib.add_book(book1)
lib.add_borrower(person1)

# Borrow book
lib.borrow_book()

# Return book
lib.return_book()
```

### 4. Restaurant Order System

This example models a basic restaurant ordering system, illustrating:

*   **Multiple Classes:** `MenuItem`, `Order`, and `Customer` interact.
*   **Data Structures:** Using lists and dictionaries to manage menu items and orders.
*   **Method Interaction:** Methods in one class call methods in another (e.g., `Order` adding `MenuItem`s).
*   **Total Calculation:** Calculating the total cost of an order.

```python
class MenuItem:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} (${self.price:.2f})"

class Order:
    def __init__(self, order_id, customer_name):
        self.order_id = order_id
        self.customer_name = customer_name
        self.items = []
        self.total_cost = 0.0
        self.is_completed = False

    def add_item(self, menu_item, quantity=1):
        if quantity > 0:
            self.items.append({"item": menu_item, "quantity": quantity})
            self.total_cost += menu_item.price * quantity
            print(f"Added {quantity}x {menu_item.name} to Order {self.order_id}.")
        else:
            print("Quantity must be positive.")

    def remove_item(self, menu_item_name, quantity=1):
        item_found = False
        for item_entry in self.items:
            if item_entry["item"].name == menu_item_name:
                if item_entry["quantity"] >= quantity:
                    item_entry["quantity"] -= quantity
                    self.total_cost -= item_entry["item"].price * quantity
                    print(f"Removed {quantity}x {menu_item_name} from Order {self.order_id}.")
                    if item_entry["quantity"] == 0:
                        self.items.remove(item_entry)
                    item_found = True
                    break
                else:
                    print(f"Not enough {menu_item_name} in order to remove {quantity}.")
                    item_found = True
                    break
        if not item_found:
            print(f"{menu_item_name} not found in Order {self.order_id}.")

    def complete_order(self):
        self.is_completed = True
        print(f"Order {self.order_id} for {self.customer_name} completed. Total: ${self.total_cost:.2f}")

    def display_order_summary(self):
        status = "(Completed)" if self.is_completed else "(Pending)"
        print(f"\n--- Order {self.order_id} for {self.customer_name} {status} ---")
        for entry in self.items:
            item = entry["item"]
            qty = entry["quantity"]
            print(f"  - {qty}x {item.name} @ ${item.price:.2f} each")
        print(f"Total: ${self.total_cost:.2f}")

# Define Menu Items
burger = MenuItem("Burger", 8.99)
fries = MenuItem("Fries", 3.49)
coke = MenuItem("Coke", 2.00)
pizza = MenuItem("Pizza", 12.50)

# Create an order
order1 = Order("ORD001", "Alice")
order1.add_item(burger, 2)
order1.add_item(fries)
order1.add_item(coke)
order1.display_order_summary()

# Modify the order
order1.remove_item("Coke")
order1.add_item(pizza)
order1.display_order_summary()

# Complete the order
order1.complete_order()

# Create another order
order2 = Order("ORD002", "Bob")
order2.add_item(pizza, 1)
order2.add_item(coke, 2)
order2.display_order_summary()
```

###5. Real-Life OOP Example: A Bank Account System

This example demonstrates how to model a bank account using a Python class. It will include:

*   **Class and Objects:** Defining a `BankAccount` class and creating multiple account objects.
*   **Attributes:** Storing account details like `account_number`, `holder_name`, and `balance`.
*   **`__init__` Method:** Initializing new bank accounts.
*   **Methods:** Defining actions like `deposit`, `withdraw`, and `display_balance`.
*   **Dynamic Attribute Management:** Using `getattr`, `setattr`, and `delattr` to interact with account properties programmatically.


```python
class BankAccount:
    def __init__(self, account_number, holder_name, initial_balance=0.0):
        self.account_number = account_number
        self.holder_name = holder_name
        self.balance = initial_balance
        print(f"Account {self.account_number} created for {self.holder_name} with initial balance ${self.balance:.2f}.")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0:
            if self.balance >= amount:
                self.balance -= amount
                print(f"Withdrew ${amount:.2f}. New balance: ${self.balance:.2f}.")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    def display_balance(self):
        print(f"Account {self.account_number} ({self.holder_name}) current balance: ${self.balance:.2f}.")


# --- Creating Bank Account Objects ---
print("\n--- Creating Accounts ---")
account1 = BankAccount("1001", "Alice Smith", 500.00)
account2 = BankAccount("1002", "Bob Johnson") # Defaults to 0.00 initial balance

# --- Demonstrating Methods ---
print("\n--- Demonstrating Methods ---")
account1.deposit(200.00)
account1.withdraw(150.00)
account1.display_balance()

account2.deposit(1000.00)
account2.withdraw(1200.00) # Should fail due to insufficient funds
account2.display_balance()

# --- Dynamic Attribute Manipulation with getattr, setattr, delattr ---
print("\n--- Dynamic Attribute Manipulation ---")

# Using getattr()
print(f"\nAccessing attributes with getattr():")
acc1_holder = getattr(account1, "holder_name")
acc1_balance = getattr(account1, "balance")
print(f"Account 1 Holder: {acc1_holder}, Balance: ${acc1_balance:.2f}")

# Trying to get a non-existent attribute with a default
acc1_type = getattr(account1, "account_type", "Savings")
print(f"Account 1 Type (using default): {acc1_type}")

# Using setattr()
print(f"\nUpdating and adding attributes with setattr():")
setattr(account1, "balance", 750.00) # Update existing attribute
print(f"Updated Account 1 Balance using setattr(): ${account1.balance:.2f}")

setattr(account2, "account_type", "Checking") # Add new attribute
print(f"Added Account 2 Type using setattr(): {account2.account_type}")

# Using delattr()
print(f"\nDeleting attributes with delattr():")
print(f"Before deletion - Account 2 has 'account_type': {hasattr(account2, 'account_type')}")
delattr(account2, "account_type")
print(f"After deletion - Account 2 has 'account_type': {hasattr(account2, 'account_type')}")
```
