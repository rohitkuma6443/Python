# Exception Handling in Python

This notebook explains exception handling step by step with simple examples.

Exception handling allows a program to continue running even when errors occur.

## 1. What is an Exception?

An exception is a runtime error that interrupts the normal flow of a program.

Examples:
- Dividing by zero
- Accessing an invalid index
- Converting text to a number incorrectly

```python
# Example of an exception
x = 10
y = 0
print(x / y)
```

## 2. try and except

The try block contains code that may cause an exception.
The except block handles the error.

```python
try:
    x = 10
    y = 0
    print(x / y)
except:
    print("An error occurred")
```

## 3. Handling Specific Exceptions

It is a good practice to catch specific exceptions instead of using a generic except.

```python
try:
    num = int("abc")
except ValueError:
    print("ValueError: Invalid conversion to integer")
```

## 4. Multiple except Blocks

You can handle different exceptions separately.

```python
try:
    a = [1, 2, 3]
    print(a[1])
except IndexError:
    print("IndexError: List index out of range")
except ValueError:
    print("ValueError occurred")
```

## 5. else Block

The else block runs only if no exception occurs.

```python
try:
    x = int(input("Enter a number: "))
    y = 2
    print(x / y)
except ValueError:
    print("ValueError occurred")
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Division successful")
```

## 6. finally Block

The finally block always runs, whether an exception occurs or not.

Used for cleanup tasks like closing files or database connections.

```python
try:
    f = open("sample.txt", "w")
    f.write("Hello Python")
except IOError:
    print("File error")
finally:
    f.close()
    print("File closed")
```

## 7. Raising Exceptions

You can manually raise exceptions using the raise keyword.

```python
try:
  age = -5
  if age < 0:
    raise ValueError("Age cannot be negative")
except ValueError as e:
  print(e)
```

## 8. Custom Exceptions

You can define your own exception classes.

```python
class NegativeNumberError(Exception):
    pass

try:
    num = -10
    if num < 0:
        raise NegativeNumberError("Negative numbers are not allowed")
except NegativeNumberError as e:
    print(e)
```

## 9. Real World Example

Exception handling is commonly used in user input validation.

```python
try:
    num = int(input("Enter a number: "))
    print(100 / num)
except ValueError:
    print("Please enter a valid number")
except ZeroDivisionError:
    print("Number cannot be zero")
else:
    print("Operation successful")
finally:
    print("Program ended")
```

## Summary

- try: code that may fail
- except: handle errors
- else: runs if no error
- finally: always runs
- raise: manually trigger exceptions

Exception handling makes programs robust and user friendly.
