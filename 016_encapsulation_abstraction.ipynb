## Encapsulation

Encapsulation in Object-Oriented Programming means bundling data (attributes) and the methods that operate on that data within a single unit (a class). It's about hiding the internal state of an object and only allowing access or modification through controlled public methods.

**Simply put:**
*   **Data Hiding:** Keep an object's internal details private (using `__` prefix for attributes).
*   **Controlled Access:** Provide public methods (like `get` and `set` functions) to interact with that private data.

This helps to protect data from accidental changes and makes your code more organized and easier to maintain.

```python
class student:
    def __init__(self,name,marks):
        self.__name = name
        self.__marks = marks

    def get_marks(self):
        return self.__marks

    def set_marks(self,new_marks):
        if new_marks >= 0:
            self.__marks =new_marks
        else:
            print("Invalid Marks")

s = student("Ravi",85)
print(s.get_marks())
s.set_marks(95)
print(s.get_marks())
print(s.set_marks(-10))
```

```python
class BankAccount:
    def __init__(self,owner,bal):
        self.__owner = owner
        self.__bal = bal

    def get_owner(self):
        return self.__owner

    def get_balance(self):
        return self.__bal

    # Public Method to deposit funds
    def deposit(self,amt):
        if amt>0:
            self.__bal += amt
            print(f"Deposited {amt}. New Balance: {self.__bal}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self,amt):
        if 0 < amt <= self.__bal:
            self.__bal -= amt
            print(f"Withdrew {amt}. New Bal {self.__bal}")
        else:
            print("Invalid withdrawal amt or insufficient fund.")

my_account = BankAccount("Rohit",1000)
print(f"Account Owner : {my_account.get_owner()}")
print(f"Account Bal: {my_account.get_balance()}")
my_account.deposit(1000)
my_account.withdraw(3000)
```

## Abstraction

Abstraction in Object-Oriented Programming is about **hiding the complex implementation details** and **showing only the essential features** of an object. It focuses on *what* an object does rather than *how* it does it.

**Simply put:**
*   **Focus on essential details:** Provide a simplified view of an object.
*   **Hide complexity:** Keep the internal workings private.

This is often achieved using **Abstract Classes** and **Abstract Methods**. An abstract class cannot be instantiated (you can't create an object directly from it), and an abstract method is declared without an implementation (it must be implemented by concrete subclasses).

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self,color):
        self.color = color

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def get_color(self):
        pass

    def get_color(self):
        return self.color

class Circle(Shape):
    def __init__(self,color,radius):
        super().__init__(color)
        self.radius = radius

    def area(self):
        return 3.14*self.radius*self.radius

class Rectangle(Shape):
    def __init__(self,color,width,height):
        super().__init__(color)
        self.width = width
        self.height = height

    def area(self):
        return self.width *self.height

my_circle = Circle("Red",5)
my_rect = Rectangle("Yellow",5,5)

print(f"Circle Color: {my_circle.get_color()}")
print(f"Circle Area: {my_circle.area()}")
```

### Abstraction vs. Encapsulation: A Simple Analogy

*   **Encapsulation (The Black Box Principle):** Think of a **remote control**.
    *   It bundles all the complex electronics inside (data/attributes) and provides a few simple buttons (methods) on the outside.
    *   You can use the buttons to change channels or volume, but you don't need to know *how* the internal circuits work.
    *   **Goal:** To hide internal complexity and protect data.

*   **Abstraction (The "What" vs. "How" Principle):** Think of the **TV's user interface** itself.
    *   You see channel numbers, volume levels, and a picture – these are the *essential features*.
    *   You don't see the complex digital signals, decoders, or display technology that actually make the picture appear.
    *   **Goal:** To show only necessary information and hide detailed implementation.

## Polymorphism

Polymorphism means “same method name, different behavior.”

You can call the same method on different objects, and each object responds in its own way.

**Simply put:**
*   **Many Forms:** Objects of different classes can be treated as objects of a common type.
*   **Method Overriding:** A method defined in a base class can be redefined (overridden) in a derived class.
*   **Method Overloading:** (Less common in Python, but present in other languages) Multiple methods with the same name but different parameters within the same class.

#Method Overriding
A child class changes how a parent class method works.

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

animals = [Dog(), Cat()]

for a in animals:
    a.sound()
```

#Method Overloading

```python
class A:
  def show(self):
    print("Welcome")

  def show(self,firstname = ''):
    print("welcome", firstname)

  def show(self,firstname = '',lastname = ''):
    print("welcome", firstname, lastname)

a = A()
a.show()
a.show("Ravi")
a.show("Ravi","Singh")
```
