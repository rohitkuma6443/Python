# Python Inheritance

**Inheritance** allows us to define a class that inherits all the methods and properties from another class.

**Parent class** is the class being inherited from, also called base class.

**Child class** is the class that inherits from another class, also called derived class.

# Types of Inheritance in Python
**Single Inheritance** – One parent, one child.<br><br>
**Multiple Inheritance** – Multiple parents, one child. <br><br>
**Multilevel Inheritance** – Child inherits from a parent, which itself inherits from another parent.<br><br>
**Hierarchical Inheritance** – Multiple children inherit from the same parent.<br><br>
**Hybrid Inheritance** – Combination of the above.

````python
class Employee:
    # Constructor for the Employee class, takes name and id as arguments
    def __init__(self,name,id):
        self.name = name
        self.id = id

    # Method to display employee details
    def showDetails(self):
        print(f"The name of employee : {self.id} is {self.name}")

# Programmer class inherits from Employee class
class Programmer(Employee):
    # Method specific to Programmer class to show default language
    def showLanguage(self):
        print("The Default Language is Python")

# Create an instance of the Employee class
e1 = Employee("Rohan Das", 400)
# Create an instance of the Programmer class, which is also an Employee
e2 = Programmer("Harry",410)

# Call showDetails method for Employee object
e1.showDetails()
# Call showDetails method for Programmer object (inherited from Employee)
e2.showDetails()
# Call showLanguage method for Programmer object
e2.showLanguage()
````

### Multiple Inheritance Example

In multiple inheritance, a child class inherits from multiple parent classes. This allows the child class to combine functionalities from all its parent classes.

````python
class Father:
    def skills_father(self):
        print("Father has gardening skills")

class Mother:
    def skills_mother(self):
        print("Mother has cooking skills")

class Child(Father, Mother):
    def skills_child(self):
        print("Child has playing skills")

# Create an instance of the Child class
c = Child()

# Access skills from Father class
c.skills_father()

# Access skills from Mother class
c.skills_mother()

# Access skill from Child class itself
c.skills_child()
````

### Multilevel Inheritance Example

In multilevel inheritance, a child class inherits from a parent class, which in turn inherits from another parent class (a grandparent). This forms a chain of inheritance.

````python
class Grandfather:
    def skills_grandfather(self):
        print("Grandfather has farming skills")

class Father(Grandfather):
    def skills_father(self):
        print("Father has engineering skills")

class Son(Father):
    def skills_son(self):
        print("Son has coding skills")

# Create an instance of the Son class
s = Son()

# Access skill from Grandfather class
s.skills_grandfather()

# Access skill from Father class
s.skills_father()

# Access skill from Son class itself
s.skills_son()
````

### Hierarchical Inheritance Example

In hierarchical inheritance, multiple child classes inherit from a single parent class. This allows different specialized classes to share common functionalities from a base class.

````python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Create instances of Dog and Cat
dog = Dog()
cat = Cat()

# Dog accessing parent's method and its own method
dog.speak()
dog.bark()

# Cat accessing parent's method and its own method
cat.speak()
cat.meow()
````

### Hybrid Inheritance Example

Hybrid inheritance is a combination of two or more types of inheritance. A common example involves combining hierarchical inheritance with multiple inheritance. For instance, a base class is inherited by multiple child classes (hierarchical), and then one of these child classes, or another class, further inherits from multiple parents (multiple inheritance).

In the example below, `Person` is a base class. `Student` and `Teacher` inherit from `Person` (hierarchical inheritance). Then, `TeachingAssistant` inherits from both `Student` and `Teacher` (multiple inheritance), creating a hybrid structure.

````python
class Person:
    def __init__(self, name, **kwargs):
        super().__init__(**kwargs) # Call next in MRO, passing any extra keyword args
        self.name = name
        print(f"Person created: {self.name}")

    def introduce(self):
        print(f"Hi, my name is {self.name}")

class Student(Person):
    def __init__(self, name, student_id, **kwargs):
        super().__init__(name=name, **kwargs) # Pass name to Person, and other kwargs up the MRO
        self.student_id = student_id
        print(f"Student created with ID: {self.student_id}")

    def study(self):
        print(f"{self.name} (ID: {self.student_id}) is studying.")

class Teacher(Person):
    def __init__(self, name, faculty_id, **kwargs):
        super().__init__(name=name, **kwargs) # Pass name to Person, and other kwargs up the MRO
        self.faculty_id = faculty_id
        print(f"Teacher created with ID: {self.faculty_id}")

    def teach(self):
        print(f"{self.name} (Faculty ID: {self.faculty_id}) is teaching.")

# Hybrid class combining Student and Teacher (multiple inheritance)
class TeachingAssistant(Student, Teacher):
    def __init__(self, name, student_id, faculty_id, courses):
        # Use super() to correctly initialize the base classes according to MRO
        # Student's __init__ will be called first, then Teacher's, then Person's
        super().__init__(name=name, student_id=student_id, faculty_id=faculty_id)
        self.courses = courses
        print(f"Teaching Assistant created for courses: {', '.join(self.courses)}")

    def assist(self):
        print(f"{self.name} is assisting in {', '.join(self.courses)}.")

# Create an instance of TeachingAssistant
ta = TeachingAssistant("Alice", "S12345", "F987", ["Math", "Physics"])

print("\n--- Demonstrating inherited methods ---")
ta.introduce() # Inherited from Person (via both Student and Teacher)
ta.study()      # Inherited from Student
ta.teach()      # Inherited from Teacher
ta.assist()     # Own method of TeachingAssistant
````

### Understanding `super()`

The `super()` function in Python is used to give you access to methods and properties of a parent or sibling class. It returns a temporary object of the superclass, which then allows you to call that superclass's methods. This is particularly useful in two main scenarios:

1.  **Calling Parent Class Constructors (`__init__`)**: When a child class overrides the `__init__` method, it often needs to call the parent class's `__init__` to ensure proper initialization of attributes defined in the parent.

    ```python
    class Parent:
        def __init__(self, name):
            self.name = name
            print(f"Parent {self.name} initialized")

    class Child(Parent):
        def __init__(self, name, age):
            super().__init__(name) # Calls Parent's __init__
            self.age = age
            print(f"Child {self.name} (age {self.age}) initialized")
    ```

2.  **Calling Parent Class Methods**: If a child class overrides a method inherited from its parent but still wants to execute the parent's version of that method, `super()` can be used.

    ```python
    class Parent:
        def greet(self):
            print("Hello from Parent")

    class Child(Parent):
        def greet(self):
            super().greet() # Calls Parent's greet method
            print("Hello from Child")
    ```

### Why `super()` is Important

*   **Maintainability**: It makes your code more maintainable by reducing redundancy. You don't have to explicitly name the parent class, which makes refactoring easier if the inheritance hierarchy changes.
*   **Method Resolution Order (MRO)**: In complex inheritance hierarchies (especially with multiple inheritance), `super()` follows the MRO, ensuring that the correct next method in the hierarchy is called. This prevents issues and promotes cooperative inheritance.
*   **Flexibility**: It allows subclasses to extend or modify the behavior of parent class methods without completely replacing them.

````python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
        print(f"Vehicle initialized: {self.brand}")

    def drive(self):
        print(f"{self.brand} is moving.")

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)  # Call parent's __init__
        self.model = model
        print(f"Car initialized: {self.brand} {self.model}")

    def drive(self):
        super().drive()  # Call parent's drive method
        print(f"The {self.brand} {self.model} is driving on the road.")

class ElectricCar(Car):
    def __init__(self, brand, model, battery_life):
        super().__init__(brand, model) # Call parent's (Car's) __init__
        self.battery_life = battery_life
        print(f"Electric Car initialized: {self.brand} {self.model} with {self.battery_life} kWh battery.")

    def charge(self):
        print(f"The {self.brand} {self.model} is charging.")

# Create instances
v = Vehicle("Generic")
v.drive()

c = Car("Toyota", "Camry")
c.drive()

ec = ElectricCar("Tesla", "Model 3", 75)
ec.drive()
ec.charge()
````

### 1. Car Manufacturing System

This example models a car manufacturing system, demonstrating how different types of cars can inherit common properties and behaviors while also having their own unique characteristics. It highlights:

*   **Inheritance:** `ElectricCar` and `GasCar` inherit from a base `Car` class.
*   **Polymorphism:** The `start_engine` method behaves differently for electric and gas cars.
*   **Class Variables:** A `total_cars_produced` tracks all cars made.
*   **Encapsulation:** Attributes like `_fuel_level` are managed internally.

````python
class Car:
    total_cars_produced = 0 # Class variable to count all cars

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        Car.total_cars_produced += 1

    def display_info(self):
        return f"{self.year} {self.make} {self.model}"

    def start_engine(self):
        raise NotImplementedError("Subclasses must implement this method")

class GasCar(Car):
    def __init__(self, make, model, year, fuel_capacity):
        super().__init__(make, model, year)
        self._fuel_level = 0 # Protected attribute
        self.fuel_capacity = fuel_capacity

    def start_engine(self):
        if self._fuel_level > 0:
            return f"{self.display_info()}: Vroom! Engine started."
        else:
            return f"{self.display_info()}: Out of gas! Can't start."

    def refuel(self, amount):
        if amount > 0:
            self._fuel_level = min(self._fuel_level + amount, self.fuel_capacity)
            print(f"{self.make} {self.model} refueled. Fuel level: {self._fuel_level}/{self.fuel_capacity}")
        else:
            print("Refuel amount must be positive.")

class ElectricCar(Car):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self._charge_level = 100 # Initially fully charged
        self.battery_capacity = battery_capacity

    def start_engine(self):
        if self._charge_level > 10:
            return f"{self.display_info()}: Whirr... Electric motor engaged."
        else:
            return f"{self.display_info()}: Low charge! Please charge battery."

    def charge(self, percentage):
        if 0 < percentage <= 100:
            self._charge_level = min(self._charge_level + percentage, 100)
            print(f"{self.make} {self.model} charged. Charge level: {self._charge_level}%")
        else:
            print("Charge percentage must be between 1 and 100.")

# Create objects
my_gas_car = GasCar("Ford", "Mustang", 2022, 15)
my_electric_car = ElectricCar("Tesla", "Model 3", 2023, 75)

print(my_gas_car.start_engine())
my_gas_car.refuel(10)
print(my_gas_car.start_engine())

print("\n")
print(my_electric_car.start_engine())
my_electric_car.charge(5)
print(my_electric_car.start_engine())

print(f"\nTotal cars produced: {Car.total_cars_produced}")
````
