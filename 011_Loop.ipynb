## 1. Basics of Loops in Python

### 1. Introduction to Loops

A **loop** is a programming structure used to **repeat a block of code multiple times** until a condition is satisfied or until all items in a sequence are processed.

Loops are useful when:

* You need to perform the same operation repeatedly.
* You need to process each item in a collection (like a list or string).
* You want to automate repetitive tasks.

Instead of writing the same code many times, a loop allows you to **execute the code automatically multiple times**.

---

### 2. Purpose of `for` Loop and `While` Loop

🔹 for loop → used when number of iterations is known (fixed count)

🔹 while loop → used when condition decides execution (unknown count)

---

### 3. Syntax of `for` Loop and `while` Loop

#### Syntax of `for` Loop

```python
for variable in iterable:
    # block of code
```

Components:

* **variable** → Temporary variable that stores the current item.
* **iterable** → A collection of items (list, tuple, string, etc.).
* **block of code** → Instructions executed for each item.

Execution flow:

1. Take the first element from the iterable.
2. Assign it to the loop variable.
3. Execute the code block.
4. Move to the next element.
5. Repeat until elements are finished.

---

#### Syntax of `while` Loop

```python
while condition:
    # block of code
```

Components:

* **condition** → Boolean expression (`True` or `False`).
* **block of code** → Runs as long as the condition is `True`.

Execution flow:

1. Check the condition.
2. If `True`, execute the code block.
3. After execution, check the condition again.
4. Repeat until the condition becomes `False`.

---

### 4. Difference Between `for` Loop and `while` Loop

| Feature               | `for` Loop                        | `while` Loop                               |
| --------------------- | --------------------------------- | ------------------------------------------ |
| Purpose               | Iterates over a sequence          | Repeats while condition is true            |
| Iteration Count       | Usually known                     | Usually unknown                            |
| Condition             | Based on iterable elements        | Based on logical condition                 |
| Risk of Infinite Loop | Very low                          | Higher if condition never becomes false    |
| Common Use            | Traversing lists, strings, ranges | Waiting for condition, unknown repetitions |


### 1. Introduction to Iteration

Iteration is the process of executing a block of code repeatedly. In Python, this is primarily achieved through `for` and `while` loops.

- **Iterable**: An object capable of returning its members one at a time (e.g., lists, strings, range).
- **Iterator**: An object representing a stream of data; it produces the next value when `next()` is called on it.

**Reasoning**:
Create code examples to demonstrate the basic syntax of for loops, while loops, and the programmatic mechanics of iterables using iter() and next().



````python
### 1. Iteration Code Examples

# --- For Loop ---
# Iterating over an 'iterable' (the list)
fruits = ['apple', 'banana', 'cherry']
print("For Loop Output:")
for fruit in fruits:
    print(f"- {fruit}")

# --- While Loop ---
# Using a counter to control flow based on a condition
count = 0
print("\nWhile Loop Output:")
while count < 3:
    print(f"Iteration {count}")
    count += 1

# --- Under the Hood: Iterators ---
# Demonstrating the 'iterator' protocol using iter() and next()
print("\nManual Iteration Output:")
# 'fruits' is an iterable; 'it' is the iterator object
it = iter(fruits)

print(next(it))  # Returns 'apple'
print(next(it))  # Returns 'banana'
print(next(it))  # Returns 'cherry'

# Calling next(it) again would raise a StopIteration exception.
````

### Iterating Over Various Data Types

Python allows for easy iteration over different data structures. While lists are common, strings and dictionaries are also frequently iterated upon:

- **Dictionaries**: By default, iterating over a dictionary yields its keys. However, we can use specific methods:
    - `.keys()`: Iterates over the keys.

## Range Function and Control Statements


### The `range()` Function and Control Statements

#### 1. The `range()` Function
The `range()` function generates a sequence of numbers, which is indispensable for `for` loops. It can take up to three arguments:
- **`start`**: The starting value of the sequence (inclusive). Defaults to 0.
- **`stop`**: The end value of the sequence (exclusive). The sequence stops before reaching this number.
- **`step`**: The increment between each number. Defaults to 1.

**Reverse Iteration**: By providing a negative `step` value (and ensuring `start` > `stop`), we can iterate backwards (e.g., `range(10, 0, -1)`).

#### 2. Loop Control Statements
These statements allow you to alter the flow of a loop based on specific conditions:
- **`break`**: Immediately terminates the loop entirely. Execution proceeds to the first statement following the loop block.
- **`continue`**: Skips the remainder of the current loop iteration and moves directly to the next cycle of the loop.
- **`pass`**: A null operation. It acts as a placeholder when a statement is syntactically required but no action is needed (often used for stubbing out functions or empty conditional blocks).

**Reasoning**:
Create a code cell to demonstrate the range function with different arguments and the behavior of break, continue, and pass control statements as per the instructions.



````python
# --- Demonstrating range() ---
print('Range(5):', list(range(5)))
print('Range(2, 8):', list(range(2, 8)))
print('Range(0, 10, 2):', list(range(0, 10, 2)))

# Reverse Iteration (Countdown)
print('\nCountdown:')
for i in range(10, 0, -1):
    print(i, end=' ')
print('Liftoff!')

# --- Loop Control Statements ---
print('\n--- Break Example (finding 7) ---')
for num in range(1, 10):
    if num == 7:
        print(f'Found {num}, breaking loop.')
        break
    print(num, end=' ')

print('\n\n--- Continue Example (skipping even numbers) ---')
for num in range(1, 6):
    if num % 2 == 0:
        continue  # Skips the print statement for even numbers
    print(f'Odd number: {num}')

print('\n--- Pass Example (placeholder) ---')
for num in range(3):
    if num == 1:
        pass  # Placeholder: logic to be implemented later
        print('Pass executed at 1 (loop continues normally)')
    else:
        print(f'Processing {num}')
````

## Nested Loops and Advanced Concepts


### Advanced Looping Techniques

Beyond basic iteration, Python offers powerful features for handling complex data structures and control flows:

- **Nested Loops**: A loop inside another loop. The inner loop completes all its iterations for every single iteration of the outer loop. Common for processing grids or matrices.
- **Loop Unpacking**: When iterating over a sequence of tuples or lists, you can 'unpack' the elements directly into multiple variables in the loop header (e.g., `for x, y in points:`).
- **The `zip()` Function**: This function aggregates elements from two or more iterables. It allows you to iterate over multiple sequences in parallel, stopping when the shortest iterable is exhausted.
- **Loop-Else Clause**: A unique Python feature where an `else` block executes only if the loop completed naturally (i.e., it was NOT terminated by a `break` statement). This is often used for search operations.

````python
print('--- 1. Nested Loops (Coordinate Grid) ---')
for x in range(3):
    for y in range(2):
        print(f'({x}, {y})', end=' ')
    print()  # Newline after inner loop completion

print('\n--- 2. Loop Unpacking (List of Tuples) ---')
points = [(10, 'A'), (20, 'B'), (30, 'C')]
for value, label in points:
    print(f'Point {label} is at {value}')

print('\n--- 3. zip() Function (Parallel Iteration) ---')
names = ['Alice', 'Bob', 'Charlie']
scores = [85, 92, 78]
for name, score in zip(names, scores):
    print(f'{name} scored {score}')

print('\n--- 4. Loop-Else Clause (Search Example) ---')
target = 7
numbers = [1, 3, 5, 8, 9]

for n in numbers:
    if n == target:
        print(f'Found {target}!')
        break
else:
    # This runs only if the loop completes without 'break'
    print(f'Target {target} was not found in the list.')
````

## Comprehensions and Generators


### Comprehensions and Generator Expressions

Comprehensions provide a concise way to create new collections from existing iterables. They are often more readable and faster than traditional `for` loops for simple transformations.

#### Types of Comprehensions:
- **List Comprehension**: `[expression for item in iterable if condition]`
- **Dictionary Comprehension**: `{key_expr: value_expr for item in iterable if condition}`
- **Set Comprehension**: `{expression for item in iterable if condition}`


````python
# 1. List Comprehension: Squares of even numbers from 0 to 9
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(f'List Comprehension (Even Squares): {even_squares}')

# 2. Dictionary Comprehension: Mapping numbers 1-5 to their cubes
num_cubes = {x: x**3 for x in range(1, 6)}
print(f'Dictionary Comprehension (Cubes): {num_cubes}')

# 3. Set Comprehension: Extracting unique vowels from a string
text = 'comprehensions are powerful'
vowels = {char for char in text if char in 'aeiou'}
print(f'Set Comprehension (Unique Vowels): {vowels}')
````

# Example For Practice

1. Prime Number Checker (Using While Loop)

````python
num = int(input("Enter a number: "))
i = 2
is_prime = True

while i <= num // 2:
    if num % i == 0:
        is_prime = False
        break
    i += 1

if is_prime and num > 1:
    print("Prime Number")
else:
    print("Not Prime")
````

2. Reverse a Number

````python
num = int(input("Enter a number: "))
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reversed number:", reverse)
````

3. Armstrong Number Check

````python
num = int(input("Enter number: "))
temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10

if sum == num:
    print("Armstrong Number")
else:
    print("Not Armstrong")
````

4. Nested While Loop Pattern

````python
i = 1

while i <= 5:
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i += 1
````

5. Guess the Number Game

````python
secret = 7
guess = 0

while guess != secret:
    guess = int(input("Guess the number: "))

    if guess < secret:
        print("Too Low")
    elif guess > secret:
        print("Too High")

print("Correct Guess!")
````

6. Fibonacci Series Using While Loop

````python
n = 10
a = 0
b = 1
count = 0

while count < n:
    print(a)
    temp = a + b
    a = b
    b = temp
    count += 1
````

### Manual Histogram (Frequency Distribution)

````python
data = [10, 20, 10, 30, 20, 10, 40]

freq = {}
i = 0

while i < len(data):
    val = data[i]

    if val not in freq:
        freq[val] = 1
    else:
        freq[val] += 1

    i += 1

print(freq)
````
