## Introduction to File Handling

File Handling is a fundamental concept in programming that allows you to create, read, write, and manage files stored on your system.


## Types of Files

There are two main types of files you'll encounter:

1.  **Text Files**: Store data as a sequence of characters, readable by humans (e.g., `.txt`, `.csv`, `.py`, `.html`, `.json`). They are typically opened with a text editor.
    *   Each line ends with a special character called EOL (End Of Line), which is the newline character `\n`.

2.  **Binary Files**: Store data in a non-human-readable format (raw bytes), often specific to an application (e.g., `.jpg`, `.mp3`, `.exe`, `.pdf`). They require specific software to be interpreted.

## File Modes

When opening a file, you specify a 'mode' which dictates how the file will be used. Here are the common modes:

| Mode | Description                                                               | File Pointer Position | If File Doesn't Exist | If File Exists                                     |
| :--- | :------------------------------------------------------------------------ | :-------------------- | :-------------------- | :------------------------------------------------- |
| `r`  | **Read** (default) - Opens a file for reading.                            | Beginning             | `FileNotFoundError`   | Reads from beginning                               |
| `w`  | **Write** - Opens a file for writing.                                     | Beginning             | Creates new           | Truncates (empties) the file                       |
| `a`  | **Append** - Opens a file for appending.                                  | End                   | Creates new           | Appends to the end                                 |
| `x`  | **Exclusive Creation** - Creates a new file and opens it for writing.     | Beginning             | Creates new           | `FileExistsError`                                  |
| `b`  | **Binary** - Used with other modes (e.g., `rb`, `wb`) for binary files.   |                       |                       |                                                    |
| `t`  | **Text** (default) - Used with other modes (e.g., `rt`, `wt`) for text files. |                       |                       |                                                    |
| `+`  | **Update** - Opens a file for both reading and writing (e.g., `r+`, `w+`, `a+`). | Varies                | Varies                | Varies (depends on `r`, `w`, or `a` base mode)     |



## Opening a File

The `open()` function is used to open a file. It returns a file object, also called a handle, which is used to read from or write to the file.

```python
file_object = open('filename', 'mode')
```

*   `filename`: The path to the file.
*   `mode`: The mode in which the file should be opened (e.g., 'r', 'w', 'a'). Defaults to 'r'.
*   `encoding`: (Optional) Specifies the encoding for text files (e.g., 'utf-8', 'latin-1'). Highly recommended for text files to avoid encoding errors.

````python
# Example: Opening a file for writing ('w')
# If 'my_file.txt' doesn't exist, it will be created.
# If it exists, its content will be truncated (emptied).
file_write = open('my_file.txt', 'w')
file_write.write('Hello, world!\n')
file_write.close()

# Example: Opening a file for reading ('r')
file_read = open('my_file.txt', 'r')
content = file_read.read()
print(f"Content of my_file.txt: {content}")
file_read.close()
````

## Closing a File

## Using `with` Statement

The `with` statement is the recommended way to handle files in Python. It ensures that the file is automatically closed, even if errors occur during file operations.

````python
# Writing to a file using 'with'
with open('best_practice.txt', 'w') as f:
    f.write('This is the best way to handle files!\n')
    f.write('The file will be automatically closed.\n')

print('Content written and file automatically closed.')

# Reading from a file using 'with'
with open('best_practice.txt', 'r') as f:
    content = f.read()
    print(f"Content of best_practice.txt:\n{content}")

print('File read and automatically closed.')
````

## Reading Files

Once a file is opened in read mode, you can use various methods to retrieve its content.

### `read()`

Reads the entire content of the file as a single string. You can also specify an optional `size` argument to read a specific number of characters/bytes.

````python
# Create a sample file for reading
with open('sample_read.txt', 'w') as f:
    f.write('Line 1: The quick brown fox\n')
    f.write('Line 2: jumps over the lazy dog.\n')
    f.write('Line 3: Hello Python!\n')

# Read the entire file
with open('sample_read.txt', 'r') as f:
    full_content = f.read()
    print("--- Full Content ---")
    print(full_content)

# Read a specific number of characters
with open('sample_read.txt', 'r') as f:
    first_chars = f.read(10) # Reads first 10 characters
    print("\n--- First 10 Characters ---")
    print(first_chars)
    remaining_content = f.read() # Reads the rest from where it left off
    print("--- Remaining Content ---")
    print(remaining_content)
````

### `readline()`

Reads one entire line from the file, including the newline character `\n` at the end (if present). Returns an empty string if the end of the file is reached.

````python
with open('sample_read.txt', 'r') as f:
    print("--- Reading line by line ---")
    line1 = f.readline()
    print(f"Line 1: {line1.strip()}") # .strip() removes leading/trailing whitespace, including newline
    line2 = f.readline()
    print(f"Line 2: {line2.strip()}")
    line3 = f.readline()
    print(f"Line 3: {line3.strip()}")
    line4 = f.readline()
    print(f"Line 4 (empty if EOF): '{line4}'")
````

````python
#This is preferred for large files because it reads one
#line at a time instead of loading everything into memory.
with open("file.txt","r") as f:
    for line in f:
        print(line.strip())
````

### `readlines()`

Reads all lines from the file and returns them as a list of strings, where each string represents a line including the newline character.

````python
with open('sample_read.txt', 'r') as f:
    all_lines = f.readlines()
    print("--- Reading all lines into a list ---")
    print(all_lines)

    print("\n--- Iterating through the list ---")
    for line in all_lines:
        print(line.strip())
````

### Iterating Through a File Object

The most memory-efficient way to read a file line by line is to iterate directly over the file object. This reads one line at a time, making it suitable for very large files.

````python
print("--- Iterating directly over file object ---")
with open('sample_read.txt', 'r') as f:
    for line_num, line in enumerate(f, 1):
        print(f"Line {line_num}: {line.strip()}")
````

## Writing Files

When a file is opened in write mode (`'w'`), existing content is truncated. If the file doesn't exist, it's created. When opened in append mode (`'a'`), new content is added to the end of the file.

### `write()`

Writes a single string to the file. It does not automatically add a newline character, so you must include `\n` if you want newlines.

````python
# Writing to a new file or overwriting an existing one
with open('output.txt', 'w') as f:
    f.write('This is the first line.\n')
    f.write('This is the second line.\n')
    f.write('And this is the last line.')

print("Content written to output.txt in 'w' mode.")

# Verify content
with open('output.txt', 'r') as f:
    print("\n--- Content of output.txt ---")
    print(f.read())
````

### `writelines()`

Writes a sequence of strings (e.g., a list of strings) to the file. Like `write()`, it does not add newlines automatically, so each string in the sequence should include `\n` if a new line is desired.

````python
lines_to_write = [
    'Item A\n',
    'Item B\n',
    'Item C\n'
]

with open('items.txt', 'w') as f:
    f.writelines(lines_to_write)

print("Content written to items.txt using writelines.")

# Verify content
with open('items.txt', 'r') as f:
    print("\n--- Content of items.txt ---")
    print(f.read())
````

## Appending Data to a File

To add new content to an existing file without deleting its previous content, use the append mode (`'a'`). If the file does not exist, it will be created.

````python
# Start with a file
with open('log.txt', 'w') as f:
    f.write('Initial log entry.\n')

print("Initial content of log.txt:")
with open('log.txt', 'r') as f:
    print(f.read())

# Append new content
with open('log.txt', 'a') as f:
    f.write('Second log entry.\n')
    f.write('Third log entry.\n')

print("\nAppended content to log.txt:")
with open('log.txt', 'r') as f:
    print(f.read())
````

## File Pointer (`tell` & `seek`)

When you interact with a file, there's an internal "file pointer" (or cursor) that keeps track of your current position within the file. You can manage this pointer using `tell()` and `seek()`.

*   **`tell()`**: Returns the current position of the file pointer (offset from the beginning of the file, in bytes).
*   **`seek(offset, whence)`**: Changes the position of the file pointer.
    *   `offset`: Number of bytes to move.
    *   `whence`: (Optional) Specifies the reference point.
        *   `0` (default): Start of the file.
        *   `1`: Current position.
        *   `2`: End of the file.

`seek` and `tell` generally work with bytes, which is straightforward for binary files. For text files opened in text mode, `seek`'s behavior can be complex due to character encodings (variable byte lengths per character), so it's generally safest to only `seek(0)` (to the beginning) or to positions returned by `tell()`.

````python
with open('pointer_example.txt', 'w+') as f:
    f.write('abcdefghijklmn')
    print(f"Initial content written. Current pointer position: {f.tell()}") # Should be 14 (length of string)

    f.seek(0) # Move to the beginning of the file
    print(f"After seeking to 0, pointer position: {f.tell()}")

    content = f.read(5) # Read 5 characters
    print(f"Read '{content}'. Current pointer position: {f.tell()}") # Should be 5

    f.seek(3, 0) # Move to 3rd byte from the beginning
    print(f"After seeking to 3 from start, pointer position: {f.tell()}")
    content = f.read(3)
    print(f"Read '{content}'. Current pointer position: {f.tell()}") # Should be 6

    f.seek(0) # Reset to beginning to show reading entire content after manipulations
    print("\n--- Final content of file ---")
    print(f.read())
````

## Checking File Existence (`os` module)

The `os` module provides functions for interacting with the operating system, including file system operations. `os.path.exists()` is useful for checking if a file or directory exists.

````python
import os

# Create a dummy file
with open('check_me.txt', 'w') as f:
    f.write('temporary content')

# Check if a file exists
file_name = 'check_me.txt'
if os.path.exists(file_name):
    print(f"'{file_name}' exists.")
else:
    print(f"'{file_name}' does NOT exist.")

# Check for a non-existent file
non_existent_file = 'non_existent.txt'
if os.path.exists(non_existent_file):
    print(f"'{non_existent_file}' exists.")
else:
    print(f"'{non_existent_file}' does NOT exist.")

# Clean up the dummy file
os.remove(file_name)
print(f"'{file_name}' removed for cleanup.")
````

## Deleting Files

To delete a file, you can use the `os.remove()` function from the `os` module. Be careful, as this operation is irreversible!

````python
import os

# Create a file to be deleted
file_to_delete = 'delete_me.txt'
with open(file_to_delete, 'w') as f:
    f.write('This file will be deleted.')

print(f"File '{file_to_delete}' created.")

# Verify existence before deleting
if os.path.exists(file_to_delete):
    os.remove(file_to_delete)
    print(f"File '{file_to_delete}' deleted successfully.")
else:
    print(f"File '{file_to_delete}' did not exist.")
````

## Working with CSV Files

CSV (Comma Separated Values) files are a common format for tabular data. Python's built-in `csv` module provides functionality to read and write CSV files. For more advanced data manipulation, the `pandas` library is often used.

### Using the `csv` module

````python
import csv

# 1. Writing to a CSV file
data = [
    ['Name', 'Age', 'City'],
    ['Alice', 30, 'New York'],
    ['Bob', 24, 'Los Angeles'],
    ['Charlie', 35, 'Chicago']
]

csv_filename = 'people.csv'

with open(csv_filename, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)

print(f"CSV file '{csv_filename}' created and written.")

# 2. Reading from a CSV file
print(f"\n--- Reading '{csv_filename}' ---")
with open(csv_filename, 'r', newline='') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
````

````python
import csv

csv_filename = "students.csv"

# Number of records
n = int(input("Enter number of students: "))

with open(csv_filename, 'w', newline='') as file:
    writer = csv.writer(file)

    # Writing header
    writer.writerow(["Name", "Age", "City"])

    # Taking input using loop
    for i in range(n):
        print(f"\nEnter details for student {i+1}:")

        name = input("Enter Name: ")
        age = input("Enter Age: ")
        city = input("Enter City: ")

        # Writing each row using loop
        writer.writerow([name, age, city])

print(f"\nData successfully written to '{csv_filename}'")
````
