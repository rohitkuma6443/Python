## Patterns with For Loop

### 1. Square Pattern (For Loop)

**Question:** Create a square pattern using a for loop.

**Logic:**

Rows = n, Columns = n

````python
n = 4
for i in range(n):
    for j in range(n):
        print("*", end=" ")
    print()
````

### 2. Right Triangle (Increasing) (For Loop)

**Question:** Create a right triangle pattern with increasing stars using a for loop.

**Logic:**

Stars increase → i + 1

````python
n = 4
for i in range(n):
    for j in range(i + 1):
        print("*", end=" ")
    print()
````

### 3. Right Triangle (Decreasing) (For Loop)

**Question:** Create a right triangle pattern with decreasing stars using a for loop.

**Logic:**

Stars decrease → n - i

````python
n = 4
for i in range(n):
    for j in range(n - i):
        print("*", end=" ")
    print()
````

### 8. Continuous Number Pattern (For Loop)

### 10. 0–1 Triangle (For Loop)

**Question:** Create a 0-1 triangle pattern using a for loop.

**Logic:**

(i + j) % 2

````python
n = 4
for i in range(n):
    for j in range(i + 1):
        if (i + j) % 2 == 0:
            print(1, end=" ")
        else:
            print(0, end=" ")
    print()
````

### 11. Pyramid Pattern (For Loop)

**Question:** Create a pyramid pattern using a for loop.

**Logic:**

Spaces = n-i-1

Stars = i+1

````python
n = 4
for i in range(n):
    for j in range(n - i - 1):
        print(" ", end="")
    for j in range(i + 1):
        print("*", end=" ")
    print()
````

### 12. Inverted Pyramid (For Loop)

**Question:** Create an inverted pyramid pattern using a for loop.

**Logic:**

Spaces increase, stars decrease

````python
n = 4
for i in range(n):
    for j in range(i):
        print(" ", end="")
    for j in range(n - i):
        print("*", end=" ")
    print()
````

### 13. Diamond Pattern (For Loop)

**Question:** Create a diamond pattern using a for loop.

**Logic:**

Pyramid + Inverted pyramid

````python
n = 3

# Upper pyramid
for i in range(n):
    for j in range(n - i - 1):
        print(" ", end="")
    for j in range(i + 1):
        print("*", end=" ")
    print()

# Lower inverted pyramid
for i in range(n - 1):
    for j in range(i + 1):
        print(" ", end="")
    for j in range(n - i - 1):
        print("*", end=" ")
    print()
````

### 14. Hollow Square (For Loop)

**Question:** Create a hollow square pattern using a for loop.

**Logic:**

Print * only on borders

````python
n = 4
for i in range(n):
    for j in range(n):
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

### 15. Pascal’s Triangle (For Loop)

**Question:** Create Pascal’s Triangle pattern using a for loop.

**Logic:**

Use formula update:
num = num * (i - j) // (j + 1)

````python
n = 5
for i in range(n):
    num = 1
    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)
    print()
````

## Grocery Bill Calculator



**Question:** Create a grocery bill calculator that sums up the prices of items in a shopping cart using a loop.

**Logic: ** Iterate through a list of item prices and add each price to a running total.

````python
item_prices = [2.50, 1.25, 3.00, 0.75, 5.00]
total_bill = 0

for price in item_prices:
    total_bill += price

print(f"Total Grocery Bill: ${total_bill:.2f}")
````

This code calculates the total cost of items in a shopping cart.
- `item_prices`: This list holds the individual prices of the items in the shopping cart.
- `total_bill = 0`: Initializes a variable to keep track of the accumulated total bill, starting at zero.
- `for price in item_prices:`: This `for` loop iterates through each `price` in the `item_prices` list.
- `total_bill += price`: Inside the loop, each `price` is added to the `total_bill` variable, updating the running total.
- `print(f"Total Grocery Bill: ${total_bill:.2f}")`: After the loop finishes, this line prints the final `total_bill`, formatted to two decimal places for currency.

## Countdown Sequence



### Question:
Create a countdown sequence from a user-defined starting number down to 1 using a loop.

### Logic:
Use a loop to iterate from a starting number down to 1. Print each number in the sequence.

````python
start_number = 5

print(f"Countdown from {start_number}:")
while start_number >= 1:
    print(start_number)
    start_number -= 1
````

This code simulates a countdown sequence.
- `start_number = 5`: Initializes the starting point of the countdown.
- `print(f"Countdown from {start_number}:")`: Prints an introductory message indicating the start of the countdown.
- `while start_number >= 1:`: This `while` loop continues as long as the value of `start_number` is greater than or equal to 1. This is the condition that controls the loop's execution.
- `print(start_number)`: Inside the loop, the current value of `start_number` is printed in each iteration.
- `start_number -= 1`: After printing, `start_number` is decremented by 1. This step is crucial for the countdown to progress and eventually meet the loop termination condition (`start_number < 1`), preventing an infinite loop.

## Vowel Counter

### Question:
Create a vowel counter that takes a word and counts the number of vowels (a, e, i, o, u) it contains using a loop.

**Logic** :Iterate through each character of the word and check if it is a vowel. Increment a counter if it is.
```

````python
word = "Programming"
vowel_count = 0
vowels = "aeiou"

for char in word:
    if char.lower() in vowels:
        vowel_count += 1

print(f"The word '{word}' contains {vowel_count} vowels.")
````

This code counts the number of vowels in a given word.
- `word = "Programming"`: Defines the input string for which vowels will be counted.
- `vowel_count = 0`: Initializes a counter variable to keep track of the number of vowels found.
- `vowels = "aeiou"`: Creates a string containing all lowercase vowels for easy comparison.
- `for char in word:`: This `for` loop iterates through each character in the `word` string.
- `if char.lower() in vowels:`: Inside the loop, `char.lower()` converts the current character to lowercase. This ensures that both uppercase (e.g., 'A') and lowercase (e.g., 'a') vowels are counted. The `in vowels` checks if the lowercase character is present in the `vowels` string.
- `vowel_count += 1`: If the character is a vowel, the `vowel_count` is incremented by 1.
- `print(f"The word '{word}' contains {vowel_count} vowels.")`: After checking all characters in the word, this line prints the final count of vowels using an f-string for formatted output.

## Simple Interest Calculation


### Question:
Calculate the total amount (principal + simple interest) for a given principal, rate, and number of years using a loop.

### Logic:
Use a loop to apply the simple interest calculation for each year and accumulate the interest to the principal.

````python
principal = 1000  # Initial principal amount
rate = 0.05       # Annual interest rate (5%)
years = 5         # Number of years

print(f"Initial Principal: ${principal:.2f}")
print(f"Annual Interest Rate: {rate * 100:.0f}%")
print(f"Number of Years: {years}")

for year in range(1, years + 1):
    simple_interest = principal * rate
    principal += simple_interest
    print(f"After year {year}: Total amount = ${principal:.2f}")

print(f"\nFinal Total Amount after {years} years: ${principal:.2f}")
````

## Average Student Score



### Question:
Calculate the average score from a predefined list of student test scores using a loop.
```

### Logic:
Iterate through the list of scores, sum them up, and then divide by the number of scores.
```

````python
student_scores = [85, 92, 78, 95, 88]
total_score = 0

for score in student_scores:
    total_score += score

average_score = total_score / len(student_scores)

print(f"Student Scores: {student_scores}")
print(f"Total Score: {total_score}")
print(f"Number of Scores: {len(student_scores)}")
print(f"Average Score: {average_score:.2f}")
````

## Find Highest Temperature


### Question:
Find the highest temperature from a predefined list of daily temperatures using a loop.


### Logic:
Initialize a variable with the first temperature as the highest. Iterate through the remaining temperatures, and if a temperature is higher than the current highest, update the highest temperature variable.


````python
daily_temperatures = [25, 28, 22, 30, 27, 29, 26]
highest_temp = daily_temperatures[0]

for temp in daily_temperatures:
    if temp > highest_temp:
        highest_temp = temp

print(f"The highest temperature is: {highest_temp}°C")
````

## Basic Inventory Check

### Question:
Generate a Python code example in a question and answer format that checks if a specific item exists in a predefined inventory list and prints whether it was found or not, using a loop.

### Logic:
Initialize a boolean flag to `False`. Iterate through each item in the inventory list. If the current item matches the target item, set the flag to `True` and `break` the loop. After the loop, check the flag to determine if the item was found.

````python
inventory = ["apple", "banana", "orange", "grape", "kiwi"]
target_item = "orange"

item_found = False
for item in inventory:
    if item == target_item:
        item_found = True
        break

if item_found:
    print(f"The item '{target_item}' was found in the inventory.")
else:
    print(f"The item '{target_item}' was not found in the inventory.")
````

## Find and Store Numbers

Iterate through numbers from 1500 to 2700. For each number, check if it is divisible by 7 and a multiple of 5. If both conditions are true, add the number to a list.


````python
result_numbers = []

for num in range(1500, 2701):
    if num % 7 == 0 and num % 5 == 0:
        result_numbers.append(num)

print("Numbers divisible by 7 and multiple of 5 between 1500 and 2700:")
print(result_numbers)
````

Write a Python program that:

Takes two inputs:

m → number of rows

n → number of columns

Creates a 2D array (list of lists)

Each element = i * j

````python
m = int(input("Enter number of rows: "))
n = int(input("Enter number of columns: "))

matrix = []

for i in range(m):
    row = []
    for j in range(n):
        value = int(input(f"Enter value for position [{i}][{j}]: "))
        row.append(value)
    matrix.append(row)

# Print matrix
print("\nMatrix is:")
for row in matrix:
    for val in row:
        print(val, end=" ")
    print()
````
