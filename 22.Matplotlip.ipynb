# What is Matplotlib?

Matplotlib is a powerful and widely used Python library for creating static, animated, and interactive visualizations in Python. Think of it as a digital canvas where you can draw various types of plots and charts to understand your data better.

# Why use it?

1. Understanding Data: Visualizing data helps in identifying patterns, trends, and outliers.
2. Communication: Great plots make it easier to explain your findings to others.
3. Customization: Matplotlib offers extensive control over every aspect of a plot.

1. Import `matplotlib.pyplot`: This is the most common module used for plotting.
 We usually alias it as `plt` for convenience.
2. Prepare your data: You need numbers to plot! These can be lists or arrays.
3. Use plotting functions: `plt.plot()` for line plots, `plt.scatter()` for scatter plots, etc.
4. Customize (optional but recommended): Add labels, titles, grid, etc.
5. Show the plot: `plt.show()` displays your masterpiece.

```python
# Step 1: Import the necessary library
import matplotlib.pyplot as plt

# Step 2: Prepare some data
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 1, 6, 3]

# Step 3: Create the plot
plt.plot(x_values, y_values) # This creates a line plot with x and y data

# Step 4: Add some helpful details (customization)
plt.title('My First Simple Line Plot') # Title for the plot
plt.xlabel('X-axis Label')           # Label for the X-axis
plt.ylabel('Y-axis Label')           # Label for the Y-axis
plt.grid(True)                       # Add a grid for readability

# Step 5: Display the plot
plt.show()

print('\nAnd there you have it! Your very first Matplotlib plot. You can experiment by changing the `x_values` and `y_values`!')
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2023,2024,2025,2026,2027])
y = np.array([15,20,30,20,20])
print(x)
print(y)

plt.plot(x,y, marker="h",   markersize = 30, #ms
         markerfacecolor = "blue",    markeredgecolor = "Red",
         linestyle = "dashdot",   linewidth= "2",  color = "#f53b44")

plt.show()
```

```python
#Program 4: Marker and Dictionary multi plot

import matplotlib.pyplot as plt
import numpy as np
x = np.array([2023,2024,2025,2026,2027])
y = np.array([15,20,30,20,20])
z = np.array([20,25,36,20,40])
print(x)
print(y)
print(z)

line =  dict(marker="h",   markersize = 30, #ms
        markerfacecolor = "blue",    markeredgecolor = "Red",
        linestyle = "dashdot",   linewidth= "2",)

plt.plot(x,y,color = "#a422f5",**line)
plt.plot(x,z,color = "#22bdf5",**line)
plt.show()
```

```python
#Program 5 :  Labels and Title
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2023,2024,2025,2026,2027])
y = np.array([15,20,30,20,20])
z = np.array([20,25,36,20,40])
print(x)
print(y)
print(z)

line =  dict(marker="h",   markersize = 30, #ms
        markerfacecolor = "blue",    markeredgecolor = "Red",
        linestyle = "dashdot",   linewidth= "2",)

plt.plot(x,y,color = "#a422f5",**line)
plt.plot(x,z,color = "#22bdf5",**line)

plt.title("class size", fontsize = 20, family = "Times New Roman", fontweight = "bold", color = "#22bdf5")
plt.xlabel("Year",fontsize = 20, family = "Arial", fontweight = "bold", color = "#2dbefc")
plt.ylabel("Studnet Marks",fontsize = 20, family = "Arial", fontweight = "bold", color = "#2dbefc")
plt.xticks(x)
plt.show()
```

```python
#Program 8 (Scater Plot)
# Import Matplotlib
import matplotlib.pyplot as plt
# Data for Class A
hours_studied_A = [1, 2, 3, 4, 5, 6, 7, 8]
test_scores_A = [35, 40, 50, 55, 65, 70, 80, 90]
# Data for Class B
hours_studied_B = [1, 2, 3, 4, 5, 6, 7, 8]
test_scores_B = [30, 38, 45, 52, 60, 68, 75, 85]

# Create Scatter Plot
plt.scatter(    hours_studied_A,    test_scores_A,    color='blue',    s=100,    alpha=0.7,    label='Class A')

plt.scatter(
    hours_studied_B,    test_scores_B,    color='red',    s=100,    alpha=0.7,    label='Class B')

# Add Labels
plt.xlabel("Hours Studied")
plt.ylabel("Test Scores")

# Add Title
plt.title("Scatter Plot: Study Hours vs Test Scores")

# Add Grid
plt.grid(True)

# Add Legend
plt.legend()

# Show Plot
plt.show()
```

```python
#Program 9 (Pie Chart)
# Import Matplotlib
import matplotlib.pyplot as plt

# Data
classes = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
students = [120, 100, 80, 60]

# Colors for each slice
colors = ["skyblue", "lightgreen", "orange", "pink"]

# Explode the Seniors slice
explode = (0, 0, 0, 0.1)

# Create Pie Chart
plt.pie(
    students,
    labels=classes,
    colors=colors,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    startangle=90
)

# Add Title
plt.title("College Student Distribution")

# Show Chart
plt.show()
```
