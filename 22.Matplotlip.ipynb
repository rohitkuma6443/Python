# What is Matplotlib?

Matplotlib is a powerful and widely used Python library for creating static, animated, and interactive visualizations in Python. Think of it as a digital canvas where you can draw various types of plots and charts to understand your data better.

# Why use it?

1. Understanding Data: Visualizing data helps in identifying patterns, trends, and outliers.
2. Communication: Great plots make it easier to explain your findings to others.
3. Customization: Matplotlib offers extensive control over every aspect of a plot.

1. Import `matplotlib.pyplot`: This is the most common module used for plotting.
 We usually alias it as `plt` for convenience.
2. Prepare your data: You need numbers to plot! These can be lists or arrays.
3. Use plotting functions: `plt.plot()` for line plots, `plt.scatter()` for scatter plots, etc.
4. Customize (optional but recommended): Add labels, titles, grid, etc.
5. Show the plot: `plt.show()` displays your masterpiece.

```python
# Step 1: Import the necessary library
import matplotlib.pyplot as plt

# Step 2: Prepare some data
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 1, 6, 3]

# Step 3: Create the plot
plt.plot(x_values, y_values) # This creates a line plot with x and y data

# Step 4: Add some helpful details (customization)
plt.title('My First Simple Line Plot') # Title for the plot
plt.xlabel('X-axis Label')           # Label for the X-axis
plt.ylabel('Y-axis Label')           # Label for the Y-axis
plt.grid(True)                       # Add a grid for readability

# Step 5: Display the plot
plt.show()

print('\nAnd there you have it! Your very first Matplotlib plot. You can experiment by changing the `x_values` and `y_values`!')
```

```python
#Program 8 (Scater Plot)
# Import Matplotlib
import matplotlib.pyplot as plt
# Data for Class A
hours_studied_A = [1, 2, 3, 4, 5, 6, 7, 8]
test_scores_A = [35, 40, 50, 55, 65, 70, 80, 90]
# Data for Class B
hours_studied_B = [1, 2, 3, 4, 5, 6, 7, 8]
test_scores_B = [30, 38, 45, 52, 60, 68, 75, 85]

# Create Scatter Plot
plt.scatter(    hours_studied_A,    test_scores_A,    color='blue',    s=100,    alpha=0.7,    label='Class A')

plt.scatter(
    hours_studied_B,    test_scores_B,    color='red',    s=100,    alpha=0.7,    label='Class B')

# Add Labels
plt.xlabel("Hours Studied")
plt.ylabel("Test Scores")

# Add Title
plt.title("Scatter Plot: Study Hours vs Test Scores")

# Add Grid
plt.grid(True)

# Add Legend
plt.legend()

# Show Plot
plt.show()
```

```python
#Program 9 (Pie Chart)
# Import Matplotlib
import matplotlib.pyplot as plt

# Data
classes = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
students = [120, 100, 80, 60]

# Colors for each slice
colors = ["skyblue", "lightgreen", "orange", "pink"]

# Explode the Seniors slice
explode = (0, 0, 0, 0.1)

# Create Pie Chart
plt.pie(
    students,
    labels=classes,
    colors=colors,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    startangle=90
)

# Add Title
plt.title("College Student Distribution")

# Show Chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1, 6)

# Create Figure and 2x2 Subplots
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1 : Line Plot
ax[0,0].plot(x, x, color='blue', marker='o')
ax[0,0].set_title("Linear Function")
ax[0,0].set_xlabel("X Values")
ax[0,0].set_ylabel("Y Values")
ax[0,0].grid(True)

# Plot 2 : Square Plot
ax[0,1].plot(x, x**2, color='red', marker='s')
ax[0,1].set_title("Square Function")
ax[0,1].set_xlabel("X Values")
ax[0,1].set_ylabel("Y Values")
ax[0,1].grid(True)

# Plot 3 : Bar Chart
subjects = ["Python", "SQL", "Power BI", "Excel"]
marks = [85, 78, 90, 88]

ax[1,0].bar(subjects, marks, color='green')
ax[1,0].set_title("Student Marks")
ax[1,0].set_xlabel("Subjects")
ax[1,0].set_ylabel("Marks")

# Plot 4 : Pie Chart
sales = [40, 30, 20, 10]
labels = ["North", "South", "East", "West"]

ax[1,1].pie(sales,
            labels=labels,
            autopct='%1.1f%%',
            shadow=True)

ax[1,1].set_title("Regional Sales")

# Adjust Layout
plt.tight_layout()

# Show Output
plt.show()
```

What is a Box Plot?

A box plot is a graphical method used to represent the distribution of a dataset using summary statistics. It helps you understand the center, spread, and variability of data in a single diagram.

It mainly shows:

Minimum value
First quartile (Q1)
Median (Q2)
Third quartile (Q3)
Maximum value
Outliers (if any)

```python
import matplotlib.pyplot as plt

# Sample data
data = [10, 12, 14, 15, 18, 20, 22, 24, 30]

# Create box plot
plt.boxplot(data)

# Title
plt.title("Basic Box Plot")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [-100,10, 12, 14, 15, 18, 20, 22, 24, 30, 100]  # 100 is outlier

plt.boxplot(
    data,
    patch_artist=True,
    boxprops=dict(facecolor="lightblue"),
    medianprops=dict(color="red", linewidth=2),
    whiskerprops=dict(color="green"),
    capprops=dict(color="black")
)

plt.title("Customized Box Plot with Outlier")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 25, 15, 30, 20]

# Line plot
plt.plot(x, y, color='blue')

# Area fill
plt.fill_between(x, y, color='skyblue', alpha=0.5)

# Title and labels
plt.title("Simple Area Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

y1 = [10, 20, 25, 30, 40]
y2 = [5, 15, 10, 20, 25]

# First area
plt.plot(x, y1, color='blue')
plt.fill_between(x, y1, color='blue', alpha=0.3, label="Series 1")

# Second area
plt.plot(x, y2, color='green')
plt.fill_between(x, y2, color='green', alpha=0.3, label="Series 2")

plt.title("Multiple Area Chart")
plt.xlabel("X Axis")
plt.ylabel("Values")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

y1 = [10, 20, 30, 40, 50]
y2 = [5, 10, 15, 20, 25]
y3 = [20, 4, 6, 8, 10]

# Stacked area
plt.stackplot(x, y1, y2,y3, labels=["A", "B"], colors=['blue', 'orange'], alpha=0.6)

plt.title("Stacked Area Chart")
plt.xlabel("X Axis")
plt.ylabel("Values")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
data = [12, 15, 18, 20, 22, 25, 27, 30, 32, 35, 37, 40, 42, 45, 48, 50]

# Create histogram and store values
counts, bins, patches = plt.hist(data, bins=5, color='green', edgecolor='black')

# Add frequency labels on each bar
for i in range(len(counts)):
    plt.text(
        (bins[i] + bins[i+1]) / 2,   # X position (center of bar)
        counts[i],                   # Y position (height of bar)
        int(counts[i]),              # Text (frequency count)
        ha='center', va='bottom'
    )

# Titles and labels
plt.title("Histogram with Frequency Count")
plt.xlabel("Value Range")
plt.ylabel("Frequency")

# Show plot
plt.show()
```
