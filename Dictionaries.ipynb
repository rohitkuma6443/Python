## Introduction to Dictionaries

A Dictionary in Python is a data structure used to store data in key–value pairs. Each key is unique and is used to access its corresponding value.

**Key Points about Dictionary**

1. Stores data as Key : Value
2. Keys must be unique
3. Values can be duplicated
4. Dictionary is mutable (you can change or update data)
5. It is written using curly brackets {}

### Creating Dictionaries

Dictionaries can be created in several ways:

1.  **Using curly braces `{}`**: This is the most common way to create a dictionary, where key-value pairs are separated by commas, and keys are separated from values by a colon.
2.  **Using the `dict()` constructor**: This can be used to create dictionaries from sequences of key-value pairs (e.g., a list of tuples) or by passing keyword arguments.

Keys within a dictionary must be unique. If you try to assign a value to an existing key, the old value will be overwritten.

````python
# 1. Using curly braces
my_dict = {'name': 'Alice', 'age': 30, 'city': 'New York'}
print("Dictionary created with curly braces:", my_dict)

# 2. Using the dict() constructor with keyword arguments
another_dict = dict(name='Bob', age=25, city='London')
print("Dictionary created with dict() (keyword args):", another_dict)

# 3. Using the dict() constructor with a list of tuples
third_dict = dict([('fruit', 'apple'), ('color', 'red'), ('quantity', 5)])
print("Dictionary created with dict() (list of tuples):", third_dict)

# Creating an empty dictionary
empty_dict = {}
print("Empty dictionary:", empty_dict)
````

### Dictionary Syntax

-   Dictionaries are enclosed in curly braces `{}`.
-   Each item is a key-value pair.
-   Keys and values are separated by a colon `:`.
-   Key-value pairs are separated by commas `,`.
-   Keys must be unique and immutable (e.g., strings, numbers, tuples).
-   Values can be of any data type and can be duplicates.

````python
example_dict = {
    'name': 'Charlie',       # String key, String value
    'age': 35,               # String key, Integer value
    1: 'one',                # Integer key, String value
    (2, 3): 'tuple_key',     # Tuple key, String value
    'is_student': False,     # String key, Boolean value
    'grades': [90, 85, 92]   # String key, List value
}

print(example_dict)
````

### Accessing Values by Key

You can access the value associated with a key by placing the key inside square brackets `[]` after the dictionary name. If the key does not exist, a `KeyError` will be raised.

Alternatively, you can use the `.get()` method, which allows you to specify a default value to return if the key is not found, instead of raising an error.

````python
person = {'name': 'David', 'age': 40, 'occupation': 'Engineer'}

# Accessing values using square brackets
print("Name:", person['name'])
print("Age:", person['age'])

# Using .get() method
print("Occupation (using .get()):", person.get('occupation'))
print("Country (using .get() with default):", person.get('country', 'Unknown'))

# This would raise a KeyError if 'country' didn't exist:
# print(person['country'])
````

### Adding Key–Value Pairs

To add a new key-value pair to an existing dictionary, simply assign a value to a new key using square bracket notation. If the key already exists, this operation will update the value associated with that key.

````python
student = {'id': 101, 'name': 'Eva'}
print("Original student dict:", student)

# Add a new key-value pair
student['major'] = 'Computer Science'
print("After adding 'major':", student)

student['gpa'] = 3.8
print("After adding 'gpa':", student)
````

### Updating Dictionary Values

To change the value associated with an existing key, you can again use square bracket notation and assign a new value. You can also use the `.update()` method to merge another dictionary or an iterable of key-value pairs into the current dictionary.

````python
product = {'name': 'Laptop', 'price': 1200, 'in_stock': True}
print("Original product dict:", product)

# Update an existing value using square brackets
product['price'] = 1150
print("After updating 'price':", product)

# Update multiple values using .update() with another dictionary
product.update({'in_stock': False, 'discount': 0.10})
print("After updating with another dict:", product)

# Update using .update() with a list of tuples
product.update([('weight_kg', 2.5), ('brand', 'XYZ')])
print("After updating with list of tuples:", product)
````

### Removing Items from Dictionary

There are several ways to remove items from a dictionary:

1.  **`del` keyword**: Removes a specific key-value pair. If the key does not exist, it raises a `KeyError`.
2.  **`.pop()` method**: Removes a specific key-value pair and returns the value. You can provide a default value to return if the key is not found, preventing a `KeyError`.
3.  **`.popitem()` method**: Removes and returns an arbitrary key-value pair (in Python 3.7+, it removes the last inserted item).
4.  **`.clear()` method**: Removes all items from the dictionary.

````python
inventory = {'apple': 10, 'banana': 20, 'orange': 15, 'grape': 5}
print("Original inventory:", inventory)

# 1. Using del to remove a specific item
del inventory['banana']
print("After 'del' banana:", inventory)

# 2. Using .pop() to remove and get value
orange_count = inventory.pop('orange')
print(f"After '.pop()' orange, count was {orange_count}:", inventory)

# .pop() with a default value
kiwi_count = inventory.pop('kiwi', 0)
print(f"Attempted to pop 'kiwi', default value returned: {kiwi_count}:", inventory)

# 3. Using .popitem() (removes last inserted in Python 3.7+)
last_item = inventory.popitem()
print(f"After '.popitem()', removed {last_item}:", inventory)

# 4. Using .clear() (cleared in the 'Clearing Dictionaries' section)
````

### Dictionary Length

The `len()` function returns the number of key-value pairs (items) in a dictionary.

````python
my_settings = {'theme': 'dark', 'font_size': 14, 'notifications': True}
print("My settings dictionary:", my_settings)
print("Number of items in my_settings:", len(my_settings))

empty_dict = {}
print("Number of items in empty_dict:", len(empty_dict))
````

### Checking if Key Exists

To check if a key exists in a dictionary, you can use the `in` keyword. This returns `True` if the key is found, and `False` otherwise.

````python
car = {'make': 'Toyota', 'model': 'Camry', 'year': 2020}

print("Does 'make' exist?", 'make' in car)
print("Does 'color' exist?", 'color' in car)

# You can also check if a key is NOT in a dictionary
print("Is 'engine' not in car?", 'engine' not in car)
````

### Dictionary Keys

The `.keys()` method returns a view object that displays a list of all the keys in the dictionary. This view object is dynamic, meaning that if the dictionary changes, the view object reflects those changes.

````python
student_info = {'name': 'Frank', 'age': 22, 'major': 'Biology', 'university': 'State Uni'}

keys_view = student_info.keys()
print("Keys view:", keys_view)

# You can convert the keys view to a list
keys_list = list(keys_view)
print("Keys as a list:", keys_list)

# Demonstrate dynamic nature
student_info['gpa'] = 3.9
print("Keys view after adding 'gpa':", keys_view)
````

### Dictionary Values

The `.values()` method returns a view object that displays a list of all the values in the dictionary. Like `.keys()`, this view is dynamic and reflects changes to the dictionary.

````python
sales_data = {'Q1': 15000, 'Q2': 18000, 'Q3': 16500, 'Q4': 20000}

values_view = sales_data.values()
print("Values view:", values_view)

# Convert to a list
values_list = list(values_view)
print("Values as a list:", values_list)

# Demonstrate dynamic nature
sales_data['Q1'] = 15500
print("Values view after updating 'Q1':", values_view)
````

### Dictionary Items

The `.items()` method returns a view object that displays a list of a dictionary's key-value tuple pairs. This view is also dynamic.

````python
user_profile = {'username': 'jdoe', 'email': 'jdoe@example.com', 'status': 'active'}

items_view = user_profile.items()
print("Items view:", items_view)

# Convert to a list of tuples
items_list = list(items_view)
print("Items as a list of tuples:", items_list)

# Iterate through items
print("\nIterating through items:")
for key, value in user_profile.items():
    print(f"{key}: {value}")
````

### Copying Dictionaries

When you assign one dictionary to another, you're not creating a new dictionary; you're just creating a new reference to the *same* dictionary. To create a true copy (a shallow copy) of a dictionary, you can use the `.copy()` method or the `dict()` constructor.

*   **Shallow Copy**: Creates a new dictionary, but the values themselves are still references to the original objects. If values are mutable (e.g., lists, other dictionaries), changes to these nested mutable objects in the copy will affect the original, and vice versa.
*   **Deep Copy**: If you need to copy nested mutable objects independently, you'll need to use the `copy.deepcopy()` function from the `copy` module.

````python
original_dict = {'a': 1, 'b': 2, 'c': [3, 4]}
print("Original dict:", original_dict)

dict1 = {"name": "Amit", "age": 25}
dict2 = dict1
dict2["age"] = 30
print(dict1)

dict1 = {"name": "Amit", "age": 25}
dict2 = dict1.copy()
dict2["age"] = 30
print(dict1)
print(dict2)

dict1 = {"name": "Amit", "age": 25}
dict2 = dict(dict1)
````

### Clearing Dictionaries

The `.clear()` method removes all key-value pairs from a dictionary, making it empty.

````python
data = {'item1': 10, 'item2': 20, 'item3': 30}
print("Original dict:", data)

data.clear()
print("After clearing dict:", data)
````

### Nested Dictionaries

A nested dictionary is a dictionary where values are themselves dictionaries. This allows for representing more complex, hierarchical data structures.

````python
students_db = {
    'student_1': {
        'name': 'Grace',
        'age': 20,
        'courses': {'math': 'A', 'physics': 'B'}
    },
    'student_2': {
        'name': 'Harry',
        'age': 21,
        'courses': {'chemistry': 'A-', 'biology': 'A'}
    }
}

print("Nested dictionary:", students_db)

# Accessing nested values
print("\nStudent 1's name:", students_db['student_1']['name'])
print("Student 2's biology grade:", students_db['student_2']['courses']['biology'])

# Adding a new course for student 1
students_db['student_1']['courses']['history'] = 'B+'
print("\nStudent 1's updated courses:", students_db['student_1']['courses'])
````

### Dictionaries with Different Data Types

Keys must be immutable (numbers, strings, tuples), but values can be of any data type, including other dictionaries, lists, sets, functions, or custom objects. This flexibility makes dictionaries very powerful for organizing diverse data.

````python
diverse_dict = {
    'string_key': 'a simple string',
    123: 45.67,                        # Integer key, float value
    (1, 2): ['list', 'as', 'value'],   # Tuple key, list value
    'boolean_flag': True,
    'nested_dict': {'sub_key': 'sub_value', 'number': 7},
    'set_value': {10, 20, 30},
    'func_value': lambda x: x * x     # Function as a value
}

print(diverse_dict)

print("\nAccessing different types of values:")
print("String value:", diverse_dict['string_key'])
print("List value:", diverse_dict[(1, 2)])
print("Nested dict value:", diverse_dict['nested_dict'])
print("Calling function value:", diverse_dict['func_value'](5))
````

### Dictionary vs List

| Feature          | Dictionary (`{}`)                                    | List (`[]`)                                         |
| :--------------- | :--------------------------------------------------- | :-------------------------------------------------- |
| **Ordering**     | Ordered (insertion order guaranteed since Python 3.7) | Ordered (elements maintain insertion order)         |
| **Indexing**     | By unique, immutable keys (e.g., strings, numbers)   | By integer indices (0-based)                        |
| **Mutable**      | Yes (can add, remove, change items)                  | Yes (can add, remove, change elements)              |
| **Heterogeneous**| Yes (keys and values can be different types)         | Yes (elements can be different types)               |
| **Access Time**  | Fast (average O(1) for lookup by key)                | Fast (average O(1) by index), Slower (O(n) for value search)|
| **Purpose**      | Storing mappings, structured data, fast lookups      | Storing ordered sequences of items, collections     |
| **Duplicate**    | Keys must be unique; values can be duplicates        | Elements can be duplicates                          |

### Dictionary vs Set

| Feature          | Dictionary (`{}`)                                    | Set (`{}`)                                        |
| :--------------- | :--------------------------------------------------- | :------------------------------------------------- |
| **Purpose**      | Stores key-value pairs (mappings)                    | Stores unique, unordered collections of items      |
| **Structure**    | `{key: value, ...}`                                  | `{item1, item2, ...}` (or `set()`)                 |
| **Elements**     | Key-value pairs; Keys must be unique & immutable     | Unique, immutable items; Duplicates discarded      |
| **Ordering**     | Ordered (insertion order guaranteed since Python 3.7) | Unordered                                          |
| **Access**       | By key (e.g., `my_dict['key']`)                      | No direct access by index or key                   |
| **Mutability**   | Mutable (can add, remove, change items)              | Mutable (can add, remove items)                    |
| **Use Cases**    | Data records, configuration, lookups                 | Membership testing, removing duplicates, set operations |

**In summary:**

*   **Dictionaries** are for when you need to associate a value with a specific key, providing a mapping from unique keys to values.
*   **Sets** are for when you need a collection of distinct items, and the order doesn't matter. They are optimized for checking membership and eliminating duplicate entries.

### Dictionary Use Cases

Dictionaries are incredibly versatile and are used in a wide variety of scenarios:

1.  **Representing Records/Objects**: Storing data for an entity where each key is an attribute name and the value is its corresponding data (e.g., `user = {'name': 'John', 'age': 30, 'email': 'john@example.com'}`).
2.  **Counting Frequencies**: Counting occurrences of items (e.g., words in a text, votes in an election).
3.  **Mapping Data**: Creating a lookup table or mapping one set of values to another (e.g., country codes to full country names, currency symbols to names).
4.  **Configuration Settings**: Storing application or user settings.
5.  **JSON Data**: Python dictionaries map directly to JSON objects, making them ideal for working with web APIs and data exchange.
6.  **Sparse Data Representation**: Storing data where most values are zero or default, only storing the non-default values.
7.  **Implementing Hash Tables/Maps**: They are the underlying data structure for hash tables, offering efficient average-case time complexity for common operations.
8.  **Passing Keyword Arguments**: In function calls, keyword arguments are often handled internally as dictionaries.

### Example 1: Product Inventory Management

This example demonstrates how a dictionary can represent a product with details like its name (string), price (float), and a list of available colors. It also includes an `if-else` condition to check if an item is in stock based on its quantity.

````python
product = {
    'name': 'Smartphone X',
    'price': 699.99,
    'colors': ['Black', 'Silver', 'Blue'],
    'in_stock': 25
}

print(f"Product Name: {product['name']}")
print(f"Price: ${product['price']}")
print(f"Available Colors: {', '.join(product['colors'])}")

# Check stock level
if product['in_stock'] > 0:
    print(f"{product['name']} is in stock! ({product['in_stock']} units available)")
else:
    print(f"{product['name']} is currently out of stock.")

# Check if a specific color is available
if 'Blue' in product['colors']:
    print("The blue version is available.")
else:
    print("Blue is not an available color for this product.")
````

### Example 2: User Profile with Location and Interests

Here, a dictionary represents a user profile. It uses a tuple for geographic coordinates (latitude, longitude), a list for multiple interests, and strings for name and email. An `if-else` block checks the user's age category.

````python
user_profile = {
    'username': 'coder_guy',
    'email': 'coder.guy@example.com',
    'location_coords': (40.7128, -74.0060), # Tuple for (latitude, longitude)
    'interests': ['programming', 'gaming', 'travel'], # List of interests
    'age': 28
}

print(f"Username: {user_profile['username']}")
print(f"Email: {user_profile['email']}")
print(f"Location (Lat, Long): {user_profile['location_coords']}")
print(f"Interests: {', '.join(user_profile['interests'])}")

# Categorize user by age
if user_profile['age'] < 18:
    print(f"{user_profile['username']} is a minor.")
elif user_profile['age'] >= 18 and user_profile['age'] < 65:
    print(f"{user_profile['username']} is an adult.")
else:
    print(f"{user_profile['username']} is a senior citizen.")
````

### Example 3: Recipe Details

This dictionary stores information about a recipe, including its name (string), preparation time (integer), and a list of ingredients (strings). An `if-else` condition helps determine if a recipe is quick or requires more time.

````python
recipe = {
    'name': 'Spaghetti Carbonara',
    'prep_time_minutes': 25,
    'ingredients': ['spaghetti', 'eggs', 'bacon', 'parmesan cheese', 'black pepper'],
    'difficulty': 'Medium'
}

print(f"Recipe Name: {recipe['name']}")
print(f"Preparation Time: {recipe['prep_time_minutes']} minutes")
print(f"Main Ingredients: {', '.join(recipe['ingredients'])}")

# Check if the recipe is quick
if recipe['prep_time_minutes'] < 30:
    print(f"{recipe['name']} is a quick recipe!")
elif recipe['prep_time_minutes'] >= 30 and recipe['prep_time_minutes'] < 60:
    print(f"{recipe['name']} takes a moderate amount of time.")
else:
    print(f"{recipe['name']} is a long recipe.")

# Check for a specific ingredient
if 'garlic' in recipe['ingredients']:
    print("This recipe includes garlic.")
else:
    print("Garlic is not listed as an ingredient.")
````

### Example 4: Book Information

This example uses a dictionary to store details about a book. It includes the title and author (strings), publication year (integer), and a tuple representing the (number of pages, genre). An `if-else` statement checks if the book is a classic based on its publication year.

````python
book = {
    'title': '1984',
    'author': 'George Orwell',
    'publication_year': 1949,
    'details': (328, 'Dystopian Fiction') # Tuple for (pages, genre)
}

print(f"Book Title: {book['title']}")
print(f"Author: {book['author']}")
print(f"Publication Year: {book['publication_year']}")
print(f"Pages: {book['details'][0]}, Genre: {book['details'][1]}")

# Check if the book is considered a classic (e.g., published before 1970)
if book['publication_year'] < 1970:
    print(f"{book['title']} is considered a classic.")
elif book['publication_year'] >= 1970 and book['publication_year'] < 2000:
    print(f"{book['title']} is a modern classic.")
else:
    print(f"{book['title']} is a contemporary book.")
````

### Example 5: Movie Details with Ratings and Cast

This dictionary stores movie information, featuring its title (string), release year (integer), a list of genres (strings), and a tuple for the (IMDb rating, Rotten Tomatoes score). An `if-else` condition evaluates the movie's overall rating.

````python
movie = {
    'title': 'The Matrix',
    'release_year': 1999,
    'genres': ['Sci-Fi', 'Action'], # List of genres
    'ratings': (8.7, 88), # Tuple for (IMDb, Rotten Tomatoes %)
    'director': 'The Wachowskis'
}

print(f"Movie Title: {movie['title']}")
print(f"Release Year: {movie['release_year']}")
print(f"Genres: {', '.join(movie['genres'])}")
print(f"IMDb Rating: {movie['ratings'][0]}, Rotten Tomatoes: {movie['ratings'][1]}%")

# Evaluate movie rating
imdb_rating = movie['ratings'][0]
rt_score = movie['ratings'][1]

if imdb_rating >= 8.5 and rt_score >= 80:
    print(f"{movie['title']} is a highly acclaimed movie!")
elif imdb_rating >= 7.0 or rt_score >= 60:
    print(f"{movie['title']} is generally well-received.")
else:
    print(f"{movie['title']} has mixed reviews.")

# Check if 'Action' is one of the genres
if 'Action' in movie['genres']:
    print("This is an action movie.")
else:
    print("This movie is not categorized as action.")
````

### Dictionary Methods Overview

Python dictionaries come with several built-in methods to perform various operations:

*   `dict.clear()`: Removes all items from the dictionary.
*   `dict.copy()`: Returns a shallow copy of the dictionary.
*   `dict.fromkeys(seq, value)`: Creates a new dictionary with keys from `seq` and values set to `value` (default `None`).
*   `dict.get(key, default)`: Returns the value for `key` if `key` is in the dictionary, else `default`.
*   `dict.items()`: Returns a new view of the dictionary's items (key-value pairs).
*   `dict.keys()`: Returns a new view of the dictionary's keys.
*   `dict.pop(key, default)`: Removes the specified `key` and returns its value. If `key` is not found, `default` is returned if provided, else a `KeyError` is raised.
*   `dict.popitem()`: Removes and returns an arbitrary (key, value) pair. In Python 3.7+, it removes the last inserted item.
*   `dict.setdefault(key, default)`: Returns the value of `key`. If `key` is not in the dictionary, insert `key` with a value of `default` and return `default`.
*   `dict.update(iterable)`: Updates the dictionary with the key-value pairs from `iterable`, overwriting existing keys.
*   `dict.values()`: Returns a new view of the dictionary's values.
